***Condiciones***

```python
array_tridimensional[array_tridimensional > 5]   #Esto sirve para seleccionar los valores  >5 y se pone dentro de [] porque sino te devuelve true en el lugar de números >5 
                                                #array([[[False, False],[False, False]],[[False,  True],[ True,  True]]])
```

```python
array_tridimensional[array_tridimensional %2 == 0]   #Esto selecciona los pares
```

```python
array_tridimensional[(array_tridimensional> 5) & (array_tridimensional %2 == 0)]    #doble condición (no se usa and sino &) y cada condicíon (cuando hay varias) se meten en parentesis 
```

```python
array_tridimensional[(array_tridimensional> 7) & (array_tridimensional %2 != 0)]   #cuando no haya nada q lo cumpla te lo devuelve vacío
```

***índices de los elementos que cumplen la condición***

```python
np.where(array_tridimensional > 5)  #Devuelve los índices de los elementos que cumplen la condición
```

```python
np.transpose(array_tridimensional)   #Mantiene el número de filas y cambia el n de matrices a n de columnas y viceversa por defecto
```

```python
np.swapaxes(array_tridimensional, 0, 2)    #la diferencia del transpose (aquí solo se pueden modificar 2 ejes)
```
