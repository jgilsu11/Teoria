Vamos a crear dos columnas precio min y max en las q solo tengan un precio se repita su valor y las que tengan 2 que s eponga el min y el max cada uno en una columna

```python
#MIGUEL

precio_zapas_df = pd.Series(precio_zapas)          #Lo pasa de lista a series
precio_zapas_df = precio_zapas_df.str.split(" a ",expand=True)   #crea 2 columnas
precio_zapas_df.columns = ["precio_min","precio_max"]         #Renombra las columnas
precio_zapas_df["precio_max"] = np.where(precio_zapas_df["precio_max"].isna(), precio_zapas_df["precio_min"], precio_zapas_df["precio_max"])    #Donde el precio max sea nulo quiero que me lo rellenes con el precio min y sino me pones el precio max que tenga
precio_zapas_df = precio_zapas_df.applymap(lambda x: x.replace(" EUR","").replace(",","."))         #el applymap lo aplica a todo el DataFrame
precio_zapas_df.head(10)
```
