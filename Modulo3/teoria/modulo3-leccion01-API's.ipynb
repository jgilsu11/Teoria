# ¿Qué es una API?


Como científicos o analistas de datos, comprender las API's (*Application Programming Interfaces*) es muy importante. Estas herramientas nos van permitir conectar diferentes aplicaciones de software y acceder a fuentes de datos externas, es decir, van a facilitar las interacciones entre diferentes sistemas. En nuestro sector, son una parte clave para adquirir datos o enriquecer los datos que ya tenemos, ya que nos van a permitir gestionar flujos de datos en tiempo real, por ejemplo. 

Pongamos una analogía con algo rutionario para entender mejor como funcionan. Imagina que estás en un restaurante. En este escenario, el restaurante es el "servidor" que tiene varios "recursos" en el menú. Tú, como cliente, tienes la opción de seleccionar un plato del menú, lo cual es similar a utilizar una aplicación o un servicio web que necesita datos o funcionalidades de otro servicio. Aquí es donde entran en juego las APIs, o Interfaces de Programación de Aplicaciones.

Ahora, piensa en la API como el camarero de nuestro restaurante. El camarero es el intermediario que comunica tus pedidos (las solicitudes) a la cocina (el sistema o base de datos) y luego te entrega lo que has pedido (la respuesta). En el mundo digital, una API funciona de manera similar: toma tus solicitudes (obtener datos) y comunica con el sistema para obtener la información o el resultado deseado.

![api](https://github.com/Hack-io-Data/Imagenes/blob/main/02-Imagenes/API/FuncionamientoAPI_v2.png?raw=true)

Existen varios tipos de API's: 

- **APIs Web:** Estas proporcionan acceso a servicios o datos basados en internet a través de HTTP. La [API de OpenWeatherMap](https://openweathermap.org/api) proporciona datos meteorológicos actuales y previsiones para diversas ubicaciones a través de HTTP.

- **APIs de Biblioteca:** Estas ofrecen interfaces de programación para utilizar bibliotecas dentro de un lenguaje de programación específico. Por ejemplo, NumPy en Python ofrece una amplia colección de funciones para operaciones matemáticas y de álgebra lineal en arreglos multidimensionales.


- **APIs del Sistema Operativo:** Estas permiten que las aplicaciones hablen con los recursos subyacentes del sistema operativo, como sistemas de archivos, interfaces de red y hardware.  La API de archivos en Windows permite interactuar con el sistema de archivos del sistema operativo, como crear, leer, escribir y eliminar archivos. Si, lo que habéis estado usando hasta ahora con vuestros ordenadores, son API's. 


- **APIs de Terceros:** Estas proporcionan servicios o funciones de entidades externas, como plataformas de redes sociales, pasarelas de pago o servicios en la nube. Por ejemplo, la [API de X](https://developer.x.com/en/docs/twitter-api) proporciona acceso a diversas funciones de la plataforma, como la obtención de tweets, la publicación de tweets, la búsqueda de usuarios y la interacción con su contenido.


Muchas APIs utilizan autenticación para asegurar que solo los usuarios autorizados puedan acceder a ellas. Técnicas como claves API, tokens OAuth o credenciales de usuario/contraseña se utilizan para autenticar a los clientes que interactúan con la API.

Como científicos o analistas de datos, usaremos las API's en distintos contextos, pero principalmente pueden simplificar nuestro trabajo: 

- **Integración de datos**: Las APIs simplifican el proceso de recopilar datos de diferentes fuentes como redes sociales, bases de datos en línea, etc., en un único lugar.

- **Extracción de datos**: Las APIs permiten la obtención automatizada de datos de sitios web, bases de datos u otras plataformas en línea.

- **Transmisión de datos en tiempo real**: Algunas APIs proporcionan datos en tiempo real, lo que permite su uso inmediato en aplicaciones como seguimiento en vivo o análisis instantáneo.

- **Enriquecimiento de datos**: Las APIs pueden mejorar tus datos agregando información adicional como geolocalización o detalles demográficos. 

- **Transformación de datos**: Las APIs ofrecen la flexibilidad necesaria para cambiar los datos a diferentes formatos o estructuras, lo que es esencial para los procesos ETL (Extraer, Transformar, Cargar), que veremos más adelante en el curso.

- **Automatización y orquestación**: Las APIs ayudan con la automatización y gestión de flujos de datos, lo que puede ahorrar tiempo y esfuerzo.

# Conceptos importartes


- **Endpoints:** Los *endpoints* son URLs específicas que se pueden acceder a través de una API para realizar operaciones particulares. Cada endpoint generalmente representa una función o un conjunto de datos específico.

- **Métodos HTTP:** Los métodos HTTP, como GET, POST, PUT y DELETE, definen las acciones que se pueden llevar a cabo en un endpoint de la API. GET se usa para obtener datos, POST para enviar datos, PUT para actualizar datos y DELETE para eliminar datos.

- **Solicitudes y Respuestas:** Las solicitudes se realizan mediante métodos HTTP para interactuar con una API. Las respuestas de la API contienen los datos solicitados o información sobre el éxito de la solicitud.

- **Autenticación:** La autenticación es el proceso que verifica la identidad del usuario o la aplicación que realiza una solicitud a la API. Puede implicar el uso de claves de API, tokens de acceso u otros mecanismos de autenticación.

- **Limitación de Tasa (Rate Limiting):** Algunas API restringen la cantidad de solicitudes que un usuario puede realizar en un determinado período de tiempo para evitar el uso excesivo y garantizar un rendimiento adecuado del servicio.

- **Formato de Datos:** Las API pueden proporcionar datos en varios formatos, como JSON o XML. Es crucial que los analistas de datos comprendan cómo procesar y analizar estos formatos, siendo JSON el formato más comúnmente recibido.

- **Documentación:** La documentación de la API es fundamental para entender cómo utilizarla correctamente. Incluye información sobre los endpoints disponibles, los parámetros de solicitud y las respuestas esperadas.



# Buenas prácticas para trabajar con API's

Trabajar con APIs abre muchas oportunidades para acceder a datos, servicios o funcionalidades. Para aprovechar estas oportunidades, es crucial seguir algunas prácticas:

- Comprender la documentación de la API: Debemos comenzar leyendo y revisando la documentación de la API. Esto te ayuda a entender los *endpoints* , parámetros, métodos de autenticación, formatos de respuesta y restricciones de uso. Leer la documentación no es una opción, es una obligación cuando hablamos de API's. 

- Seleccionar la API correcta: Debemos elegir cuidadosamente la API que mejor se adapte a nuestras necesidades. Es importante tener en cuenta la facilidad de uso de la API, si es confiable o no, que información nos va a aportar, etc. 

- Seguir protocolos de autenticación: Después de elegir una API, debemos asegurarnos de entender y respetar su proceso de autenticación. Ya sea que implique claves de API, tokens de OAuth o autenticación básica, la autenticación correcta es clave para acceder a los puntos finales de manera segura.

- Gestionar errores de manera efectiva: Las APIs pueden devolver errores por diferentes motivos. Verifica si tu aplicación maneja estos errores de manera fluida y proporciona retroalimentación clara y útil a los usuarios, lo que se traduce en que la API tenga buena documentación, con ejemplos de uso, los errores que nos devuelve, etc. 

- Seguir los límites de llamada: Debemos respetar los límites de llamada establecidos por las APIs. Están diseñados para prevenir un uso incorrecto. Seguir estos límites ayuda a evitar ser bloqueado o limitado por el proveedor de la API.

- Mantenerse al día con la versionado de la API: Debemos realiza un seguimiento de los cambios en la versión de la API. Las APIs evolucionan, cambian, introduciendo nuevas características o eliminando las obsoletas. Siempre intentaremos usar la última versión para tener los datos más actualizados. 

# Buenas prácticas para guardar las API's keys

Como ya hemos comentado, algunas API's usan contraseñas, tokens, o *API's keys*. Mantener seguras estas claves es obligatorio, ya que algunas de las API's son de pago. Algunas recomendaciones son: 

- Usa variables de entorno: Tenemos que almacenar las claves de API en variables de entorno en lugar de incrustarlas directamente en tu código. Esto reduce el riesgo de exponer nuestras contraseñas.

- Crea un archivo .env: Una buena opción para usar variables de entorno son los archivos `.env`. Para esto, podemos usar librerías como `python-dotenv`. 

- Evita incrustar claves de API: Jamás debemos dejar las claves de API directamente en nuestro código, especialmente si planeamos compartir o hacer público nuestro código.

**Fase 3: Manejar la respuesta** 

La siguiente fase es manejar la respuesta. Cuando trabajamos con APIs, manejar la respuesta es esencial, es decir, extraer la información que realmente es valiosa para nosotros. Llegados a este punto ya hemos verificando si la llamada a la API se realizó correctamente. Es el momento de explorar y conocer en detalle toda la información que nos devuelve la API. Como lo que nos devuelve una API normalmente en un JSON, deberemos tratarlo como un diccionario e ir explorando sus claves y valores para quedarnos con la información que realemente nos interesa. Hagámoslo con nuestros datos!

Hemos visto que tenemos 50 elementos en la lista que nos devuelve la API, y podríamos intuir que cada uno de los elementos corresponde con una cervecería. A continuación, vemos un esquema de las claves que tienen cada uno de los diccionarios que tenemos:

````python
# lo primero que debemos hacer es saber qur tipo de dato tenemos en la variable 'datos_cervezas' 
print(f"El tipo de dato que tenemos en la respuesta de la API es {type(datos_cervezas)}")

# ya sabemos que es una lista, veamos que longitud tiene. En este caso, vemos que tenemos 50 resultados. 
print(f"La longitud de la lista devuelta de la API es {len(datos_cervezas)}")

# creamos un diccionario para almacenar toda la información que realmente nos interesa
diccionario_cervecerias = {"nombre": [], 
                           "tipo": [], 
                           "ciudad": [], 
                           "estado": [], 
                           "direccion": []}

# definimos un diccionario donde tenemos las claves del diccionario de la API y las claves de nuestro diccionario
campos = {"name": "nombre", 
          "brewery_type": "tipo", 
          "city": "ciudad", 
          "state": "estado", 
          "street": "direccion"}

# ya sabemos que tenemos 50 resultados, lo siguiente es entender a que corresponde cada uno de los elementos de la lista devuelta por la API
# para eso lo que vamos a hacer es hacer un for loop
for elemento in datos_cervezas:

    # mostramos el primer elemento, que es un diccionario
    print(elemento)

    # lo siguiente que hacemos es mostrar sus claves para saber que información tiene el diccionario
    print(f"Las claves del diccionario son: {elemento.keys()}")
    
    # iteramos por el diccionario de los campos de la API y de nuestro diccionario
    for campo_api, campo_diccionario in campos.items():

        # añadimos a nuestro diccionario, la información que extraemos de los datos de la API
        diccionario_cervecerias[campo_diccionario].append(elemento[campo_api])

    # ponemos un break para que solo nos muestre el primer elemento ahora que estamos explorando la información que nos devuelve la API
    break

print(f"El diccionario obtenido del primer elememto de la lista de resultados es {diccionario_cervecerias}")

# solo nos queda convertilo en DataFrame para luego poder trabajar con los datos
df_cervecerias = pd.DataFrame(diccionario_cervecerias)

# mostramos el DataFrame
df_cervecerias
````

````python
# ya lo hemos hecho para uno de los diccionarios que teníamos en la lista, ahora nos toca hacerlo para todos, lo único que tendremos que hacer es eliminar el break de nuestro código
print(f"El tipo de dato que tenemos en la respuesta de la API es {type(datos_cervezas)}")

# ya sabemos que es una lista, veamos que longitud tiene. En este caso, vemos que tenemos 50 resultados. 
print(f"La longitud de la lista devuelta de la API es {len(datos_cervezas)}")

# creamos un diccionario para almacenar toda la información que realmente nos interesa
diccionario_cervecerias = {"nombre": [], 
                           "tipo": [], 
                           "ciudad": [], 
                           "estado": [], 
                           "direccion": []}

# definimos un diccionario donde tenemos las claves del diccionario de la API y las claves de nuestro diccionario
campos = {"name": "nombre", 
          "brewery_type": "tipo", 
          "city": "ciudad", 
          "state": "estado", 
          "street": "direccion"}

# quitamos el break y lo hacemos para todos los elementos de la respuesta de la API
for elemento in datos_cervezas:

    # iteramos por el diccionario de los campos de la API y de nuestro diccionario
    for campo_api, campo_diccionario in campos.items():

        # añadimos a nuestro diccionario, la información que extraemos de los datos de la API
        diccionario_cervecerias[campo_diccionario].append(elemento[campo_api])


# solo nos queda convertilo en DataFrame para luego poder trabajar con los datos
df_cervecerias1 = pd.DataFrame(diccionario_cervecerias)

# mostramos el DataFrame
df_cervecerias1.head()
````

**Fase 4: Manejo de errores** 

Ya tenemos la llamada a un *endpoint* hecha y hemos sido capaces de gestionar la respuesta. Ahora, queda la fase de manejo de errores, es decir, incluir en nuestro código excepciones por si tenemos problemas a la hora de hacer la llamada a la API y con la limpieza de los datos. 

Dependiendo de la respuesta, nuestro código podría requerir diferentes acciones, así que deberás utilizar lógica condicional para manejar estas variaciones de forma segura, especialmente si se manejan datos sensibles, asegurándote de cumplir con las regulaciones de privacidad pertinentes (pero cuidado, no siempre será igual, cada API es un mundo y tendremos que hacerlo de forma diferente).

En este caso, gestionaremos los errores en dos puntos: 

- En la llamada a la API: En este caso, podemos tener un error al hacer la llamada a la API, algunos de los errores mque nos podemos encontrar: 

    -  **Errores del cliente (4xx)**: Que pueden incluir url mas definidas, falta de autenticación, que no tenemos permisos o que simplemente no se ha encontrado lo que esta pidiendo el usuario. 

    - **Errores del Servidor (5xx)**: Pueden incluir error genera del servidor, respuesta no valida, servidor no disponible temporalmente. 

   - **Errores de Conexión**: Incluyen problemas de red, o que el servidor no acepta conexiones. 

   -  **Errores de Tiempo de Espera (Timeout)**: Esto puede deberse a que tarda demasiado en conectarse o que tarda demasiado en recibir una respuesta. 

- En la extracción de la información: En esta situación, al estar trabajando con diccionarios, puede que tengamos problemas, porque una clave que estamos buscando no existe en la respuesta de la API, por ejemplo. 



````python
# empecemos manejando los errores de la llamada a la API

# como antes empezamos definiendo el endpoint al que vamos a hacer la llamada
url = "https://api.openbrewerydb.org/breweries"

# lo siguiente que hacemos es intentar hacer la llamada a la API, y convertir los resultados a JSON. 
# en caso de que no se puedan convertir los resultados a JSON, se irán ejecutando los except que tenemos a continuación
try:
    # Hacemos la request para extraer los datos
    response = requests.get(url)

    # Convertimos la respuesta de la API a JSON
    datos = response.json()

    # ponemos mensaje de que la conexión se realizó con éxito
    print("La conexión se realizó con éxito")

# capturamos errores HTTP, como códigos de estado 4xx (cliente) o 5xx (servidor).
except requests.exceptions.HTTPError as err:
    print(f"HTTP error: {err}")

# capturamos errores de conexión, por ejemplo, si no se puede resolver el nombre del servidor
# o si la conexión es rechazada.
except requests.exceptions.ConnectionError as err:
    print(f"Error de conexión: {err}")

# capturamos errores de tiempo de espera, como cuando la solicitud supera el tiempo límite establecido.
except requests.exceptions.Timeout as err:
    print(f"Timeout error: {err}")

# capturamos cualquier otro error relacionado con la solicitud, incluyendo problemas de red,
# errores de URL mal formateada, etc.
except requests.exceptions.RequestException as err:
    print(f"Error en la solicitud: {err}")
````

````python
# vamos a gestionar ahora los errores que pudieramos tener cuando extraemos los datos
diccionario_cervecerias = {"nombre": [], "tipo": [], "ciudad": [], "estado": [], "direccion": []}

# Definir los campos que se van a extraer
campos = {"name": "nombre", "brewery_type": "tipo", "city": "ciudad", "state": "estado", "street": "direccion"}

# Agregar datos al diccionario
for elemento in datos_cervezas:
    for campo_api, campo_diccionario in campos.items():
        # intenta acceder a la clave de la respuesta a la API
        try:
            diccionario_cervecerias[campo_diccionario].append(elemento[campo_api])

        # en caso de que la API no exista en el elemento por el que estamos iterando, añade un nulo en el diccionario
        except KeyError:
            diccionario_cervecerias[campo_diccionario].append(np.nan)

# convertimos el diccionario a DataFrame
df_cervecerias2 = pd.DataFrame(diccionario_cervecerias)
df_cervecerias2.sample(5)
````

La url de la API donde tendremos que hacer las requests es: 

```python
url = "http://www.omdbapi.com/?apikey=[yourkey]"
```

Lo primero que tendremos que hacer es conseguir nuestra API key, para ello tendremos que irnos a [esta](https://www.omdbapi.com/apikey.aspx) página y seguir los pasos que nos indican (deberéis elegir la opción gratuita). 

Los principales *endpoints* que tenemos en esta API son: 

- Búsqueda por Título:  Busca información detallada sobre una película, serie de televisión o episodio específico por su título. Recibe dos parámetros: 

    - `t`: El título de la película, serie o episodio.

    - `apikey`: Tu clave de API.


- Búsqueda por ID o IMDb ID: Busca información detallada usando el ID de IMDb. de nuevo, recibe dos parámetros: 

    - `i`: El ID de IMDb de la película, serie o episodio.

    - `apikey`: Tu clave de API.


- Búsqueda por Palabra Clave: Busca películas, series de televisión o episodios que coincidan con una palabra clave o frase. Recibe dos parámetros:

    - `s`: La palabra clave para buscar.

    - `apikey`: Tu clave de API.


- Búsqueda por Tipo: Filtra los resultados de búsqueda por tipo (movie, series, episode). Recibe tres parámetros:

    - `t`: El título de la película, serie o episodio.

    - `type`: El tipo de resultado (movie, series, episode).

    - `apikey`: Tu clave de API.



- Búsqueda por Año: Filtra los resultados de búsqueda por año. Recibe tres parámetros: 

    - `t`: El título de la película, serie o episodio.

    - `y`: El año de lanzamiento.

    - `apikey`: Tu clave de API.

- Búsqueda por Trama (Plot): Devuelve una trama completa o breve. Recibe tres parámetros: 

    - `t`: El título de la película, serie o episodio.

    - `plot`: La longitud de la trama (short o full).

    - `apikey`: Tu clave de API.


- Búsqueda por Temporada y Episodio: Devuelve información sobre una temporada y un episodio específicos. Recibe cuatros parámetros:

    - `t`: El título de la serie.

    - `Season`: El número de la temporada.

    - `Episode`: El número del episodio.

    - `apikey`: Tu clave de API.

## Archivo `.env`

Como ya hemos dicho, cuando trabajampos con APIs y necesitamos ocultar el token de acceso u otras credenciales sensibles, es una buena práctica utilizar un archivo `.env` (archivo de entorno). Veamos como hacerlo paso a paso:

1. Crear el archivo `.env`: Primero, debemos crear un archivo llamado `.env` en el directorio raíz del proyecto (recomendaríamos ponerlo a la misma altura que el jupyter notebook que estéis trabajando). Este archivo contendrá nuestras variables de entorno.

2. Agregar tus credenciales al archivo `.env`: En el archivo `.env`, debemos agregar nuestras credenciales en formato clave-valor. Por ejemplo, si tienes un token para la API de TMDB, podrías agregar algo como esto:

    ```plaintext
    tmdb_token='tu_token_aqui'
    ```

3. Instalar la librería `python-dotenv`: Para poder cargar las variables de entorno desde el archivo `.env` en nuestro código, necesitamos instalar la librería `python-dotenv`. Podremos hacerlo usando `pip` o `pip3`:

    ```sh
    pip install python-dotenv
    pip3 install python-dotenv
    ```

4. Cargar las variables de entorno en nuestro código: En nuestro jupyter, usaremos la librería `python-dotenv` para cargar las variables de entorno desde el archivo `.env`. Lo que tendríamos que hacer es:

    ```python
    import os
    from dotenv import load_dotenv

    # Cargar las variables de entorno desde el archivo .env
    load_dotenv()

    # Obtener el token de la API de las variables de entorno
    tmdb_api_token = os.getenv('tmdb_token')

    # Ahora puedes usar tmdb_api_token en tu código para hacer solicitudes a la API
    # este print no debemos dejarlo en el jupyter, porque quedaría a la vista de todo el mundo,pero si debemos hacerlo para confirmar que se cargó correctamente
    print(f'Tu token de TMDB es: {tmdb_api_token}')
    ```

5. Añadir el archivo `.env` al archivo `.gitignore`: Para asegurarnos de que el archivo `.env` no se suba a nuestro repositorio de Git y así mantener nuestras credenciales seguras, añade `.env` a tu archivo `.gitignore`:

    ```plaintext
    # .gitignore
    *.env
    ```
