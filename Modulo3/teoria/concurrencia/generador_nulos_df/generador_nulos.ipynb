# Generador de nulos para df

```python
# import panda 
import pandas as pd
import numpy as np
import os
```

```python

def editor_df(archivo):
    df = pd.read_csv(archivo,sep=';',encoding='latin1')

    print(df.info())
    print("##############################################")

    # sobreescribimos  el 25 % de los registros de la columna ANO EXERCÍCIO  
    df.loc[df.sample(frac=0.25).index, 'ANO EXERCÍCIO'] = np.nan
    df.loc[df.sample(frac=0.35).index, 'NOME ÓRGÃO SUPERIOR'] = np.nan
    df.loc[df.sample(frac=0.05).index, 'VALOR PREVISTO ATUALIZADO'] = np.nan

    print(df.info())
    print("##############################################")

    # criba de las columnas que no se les va a aplicar mas ruido 
    lista_col_eliminada = ['ANO EXERCÍCIO', 'NOME ÓRGÃO SUPERIOR','VALOR PREVISTO ATUALIZADO'] 
    lista_col= df.columns.tolist()
    for i in lista_col_eliminada:
        lista_col.remove(i)

    for i in lista_col:
        # generamos un valor aleaotiro  entre 0.05 y 0.01 y lo almacenamos en la variable proporcion_ruido
        proporcion_ruido = np.random.uniform(0.01,0.05)
        df.loc[df.sample(frac=proporcion_ruido).index, i] = np.nan

    print(df.info())
    print("##############################################")
    df.to_csv(f'datos_finales/{archivo}',sep=';',index=False)
    return df

```

```python
%%time 
lista_csv = os.listdir()
lista_csv = [i for i in lista_csv if i.endswith('.csv')]
lista_csv.sort()

# aplicamos con un map la funcion editor_df a cada uno de los archivos de la lista lista_csv
list(map(editor_df,lista_csv[:-1]))
```
