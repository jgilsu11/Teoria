# ¿Por qué es importante la visualización?

La visualización es importante para los analistas de datos por varias razones:

1. **Comprensión de los datos**: Las visualizaciones permiten explorar y comprender los datos de manera más intuitiva. Al representar los datos gráficamente, es más fácil identificar patrones, tendencias, relaciones y anomalías en los datos que pueden no ser evidentes en una tabla o conjunto de números.

2. **Identificación de insights**: Las visualizaciones pueden revelar insights valiosos ocultos en los datos. Al representar los datos en diferentes formatos gráficos, es posible descubrir patrones, correlaciones o discrepancias que pueden conducir a nuevos conocimientos y oportunidades para el negocio.

3. **Detección de errores y valores atípicos**: Las visualizaciones pueden ayudar a identificar errores en los datos o valores atípicos que pueden afectar los análisis. Al observar los gráficos, es más probable que se detecten discrepancias o valores inusuales que requieran una revisión adicional.

4. **Soporte para la toma de decisiones**: Las visualizaciones brindan una base sólida para la toma de decisiones fundamentada en datos. Al presentar información de manera visual, los analistas pueden respaldar sus argumentos con evidencia visual y facilitar la comprensión y aceptación de sus conclusiones por parte de los tomadores de decisiones.

5. **Comunicación efectiva**: Las visualizaciones ayudan a comunicar los resultados y hallazgos de análisis de datos de manera clara y concisa. Los gráficos y las visualizaciones permiten transmitir información compleja de una manera más accesible y comprensible para las partes interesadas, lo que facilita la toma de decisiones informadas.


# Countplot 

Es una visualización que muestra la frecuencia de observaciones en distintas categorías, útil para comparar la ocurrencia de cada categoría.

La estructura básica del `countplot` de Seaborn es:

```python
sns.countplot(x, data, palette=None, color=None, hue=None, orient='vertical', order=None, hue_order=None)

```

- `x`: Nombre de la columna con las categorías a contar.
  
- `data`: DataFrame o conjunto de datos en uso.
  
- `palette` (opcional): Especifica la paleta de colores del gráfico.
  
- `color`(opcional): Color de las barras.
  
- `hue` (opcional): Agrega otra variable categórica para diferenciar barras.
  
- `order` y `hue_order` (opcional): Define el orden de las categorías.
  
- `orient` (opcional): Cambia la orientación de las barras (horizontal o vertical).

Antes de empezar vamos aprender el parámetro `plt.xticks()`. Este método se utiliza para establecer o personalizar las ubicaciones y etiquetas de las marcas en el eje x de un gráfico. 

Los parámetros más importantes del método `plt.xticks()` son:

- `ticks`: Este parámetro especifica las ubicaciones en el eje x donde se colocarán las marcas. Puede ser una lista de valores numéricos que representen las posiciones de las marcas.

- `labels`: Este parámetro se utiliza para definir las etiquetas que se mostrarán en las marcas del eje x. Puede ser una lista de cadenas que representen las etiquetas de las marcas.

- `rotation`: Este parámetro permite rotar las etiquetas de las marcas en el eje x. Se especifica el ángulo de rotación en grados.

- `fontsize`: Permite ajustar el tamaño de fuente de las etiquetas en las marcas del eje x.

- `ha` (horizontal alignment): Especifica la alineación horizontal de las etiquetas en las marcas del eje x. Puede tomar los siguientes valores: 
  - `'left'`.
  
  - `'center'`.
  
  - `'right'`.

- `color`: Permite especificar el color de las etiquetas en las marcas del eje x.


Igual que tenemos este parámetro tenemos el `plt.yticks()` que recibe los mismos parámetros, y su funcionalidad es la misma solo que en el eje y.  

````python
# creamos un countplot, donde el método nos va a contar automáticamente el número de 
# filas que tenemos de cada una de las carégorias que tenemos en la columna objetivo
sns.countplot(x = "Attrition", 
              data = df, 
              palette=["#33B5FF", "#FFBB33"], 
              hue = "Attrition")

# eliminar el eje de la derecha de la gráfica
plt.gca().spines["right"].set_visible(False)

# eliminar el eje de arriba de la gráfica
plt.gca().spines["top"].set_visible(False)

# cambiar el nombre del eje y
plt.ylabel("Número de empleados")

# para cambiar el nombre de las etiquetas del eje y
plt.xticks(ticks = ["Yes", "No"], labels = ["Si", "No"]);
````

# Barplot

Es una visualización que representa datos categóricos utilizando barras rectangulares. Es una herramienta efectiva para mostrar la distribución, comparación y relaciones entre diferentes categorías. Cada barra representa una categoría y su altura está proporcional a la frecuencia, el recuento o el valor asociado con esa categoría.


## Barplot con Seaborn

La sintaxis básica de este tipo de gráfica en Seaborn es:

```python
sns.barplot(x, y, palette=None, ci='sd', hue=None)
```

Donde:

- `x`: Este parámetro especifica los valores que se ubicarán en el eje X del gráfico. En este caso, `categorias` es una lista de etiquetas o valores categóricos que se utilizarán en el eje X del gráfico de barras.

- `y`: Este parámetro establece los valores que se mostrarán en el eje Y del gráfico. Corresponde a los valores numéricos asociados con cada categoría en el eje X.

- `palette` (opcional): Aquí definimos la paleta de colores que se utilizará para colorear las barras en el gráfico. Seaborn proporciona varias paletas predefinidas, y `"Set2"` es una de ellas. Puedes utilizar otras paletas predefinidas o incluso crear tus propias paletas personalizadas.

- `ci` (opcional): Este parámetro controla la visualización del intervalo de confianza alrededor de las barras. Al establecerlo en `None`, se desactiva la representación del intervalo de confianza.

- `hue` (opcional):Permite agregar una dimensión adicional al gráfico de barras al diferenciar las barras por un factor categórico adicional.

<strong>
La mayoría de empleados permanecen en la empresa según la gráfica previa. Ahora evaluaremos si aquellos que se quedaron están más satisfechos laboralmente que los que se fueron. Se compararán porcentajes en lugar de valores absolutos, y se utilizará un subplot para presentar los resultados de manera efectiva.  </strong>
<div>

````python
# las columnas sobre las que vamos a hacer el groupby
columnas_interes = ["EnvironmentSatisfaction", "JobSatisfaction"]

# creamos un diccionario para almacenar los resultados
resultados = {}

# aplicamos la función para calcular el % de empleados por cada una de los niveles de satisfacción
for col in columnas_interes:
    resultados[col] = sv.calcular_porcentaje(df, col, "Attrition")

# mostramos los resultados
resultados.keys()
````

````python
# mostremos uno de los DataFrames que tenemos del paso anterior
resultados["EnvironmentSatisfaction"].head(2)
````

En esta gráfica vamos a añadir un método nuevo que es el `plt.legend()`. Se utiliza para mostrar/configurar la leyenda en una figura. 

Los parámetros más importantes que puede recibir son:

- `loc`: Este parámetro especifica la ubicación de la leyenda en la figura. Puede tomar valores como:       
  - `'upper right'`.
  
  - `'upper left'`.

  - `'lower right'`.
  
  - `'lower left'`.
  
  - `'center'`.
  
  - `'center right'`.
  
  - `'center left'`.
  
  - `'best'`.

- `bbox_to_anchor`: Este parámetro permite ajustar la posición de la leyenda en relación con la figura. Es una tupla que especifica las coordenadas del rectángulo de la leyenda en relación con el área de trazado.

- `fontsize`: Controla el tamaño de fuente de las etiquetas en la leyenda. Puede ser un valor numérico que represente el tamaño de fuente en puntos o una cadena que describa el tamaño de fuente relativo (por ejemplo, `'small'`, `'medium'`, `'large'`).


````python
# creamos los subplots 
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (15, 5))

# creemos el primer gráfico usando el método `barplot`
sns.barplot(x="EnvironmentSatisfaction", 
            y="EmployeeId",
            hue="Attrition",  # Variable a diferenciar por color
            data=resultados["EnvironmentSatisfaction"],  # Datos a utilizar
            palette=["#33B5FF", "#FFBB33"],  # Colores de las barras para Attrition
            order=["Low", "Medium", "High", "Very High"], # Orden de las barras en el eje x
            ax = axes[0])
# Título del primer gráfico de barras
axes[0].set_title('Satisfacción con el Ambiente de Trabajo')

# Etiqueta del eje x del primer gráfico de barras
axes[0].set_xlabel('')

# Etiqueta del eje y del primer gráfico de barras
axes[0].set_ylabel('Número de Empleados')

# Para eliminar la leyenda del primer gráfico
axes[0].legend().remove()

# eliminar la leyenda
axes[0].legend().remove()

# Graficar el segundo gráfico de barras para JobSatisfaction
sns.barplot(x="JobSatisfaction", 
            y="EmployeeId",
            hue="Attrition",  # Variable a diferenciar por color
            data=resultados["JobSatisfaction"],  # Datos a utilizar
            palette=["#33B5FF", "#FFBB33"],  # Colores de las barras para Attrition
            order=["Low", "Medium", "High", "Very High"], # Orden de las barras en el eje x
            ax = axes[1]) 

# Título del segundo gráfico de barras
axes[1].set_title('Satisfacción con el Trabajo')

# Etiqueta del eje x del segundo gráfico de barras
axes[1].set_xlabel('')

# Etiqueta del eje y del segundo gráfico de barras
axes[1].set_ylabel('Número de Empleados')

# Leyenda del segundo gráfico de barras
axes[1].legend(title="Attrition", title_fontsize='13', loc = "upper right", bbox_to_anchor = (1.20, 1));
````

<div style="background-color: #FFBB33; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>¿Cómo podemos interpretar estas gráficas?</strong>

En esta evaluación, observamos lo siguiente:

- La mayoría de los empleados optan por quedarse en la empresa en comparación con los que se van.

- Existe una correlación entre los niveles de satisfacción laboral y la retención de empleados. La mayoría de las salidas ocurren cuando los empleados reportan niveles bajos de satisfacción tanto con su trabajo como con el ambiente laboral.

- Además, se percibe una tendencia: a medida que aumenta el nivel de satisfacción, aumenta la proporción de empleados que eligen permanecer en la empresa.

## Barplot con Matplotlib

```python
plt.bar(categorias, height, data=None, color=None, hue = None)
```

Donde cada parámetro significa:

- `categorias`:  Representa las categorías o etiquetas en el eje X del gráfico de barras.

- `height`: Corresponde a los valores numéricos asociados con cada categoría en el eje X.

- `color` (opcional): Controla el color de las barras en el gráfico. 

- `data`: Es el DataFrame del que sacaremos los datos. 

- `hue`(opcional): Permite agregar una dimensión adicional a la visualización al diferenciar las barras por un factor categórico adicional.

<div style="background-color: #40B5AD; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>
En esta visualización, analizaremos la satisfacción laboral de los empleados que optaron por permanecer en comparación con los que dejaron la empresa. Se presentarán los porcentajes en lugar de valores absolutos, lo que nos permitirá identificar patrones y tendencias en la retención de empleados. Se utilizará la función groupby para este propósito. </strong>
<div>


````python
# creamos los subplots 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# establecemos el orden que queremos en el eje x
orden = ["Low", "Medium", "High", "Very High"]

# para poner un título a la gráfica
plt.suptitle("Relación entre el Nivel de Satisfacción y los empleados que se han ido")

# creemos el primer gráfico usando el método `bar`
axes[0].bar(x=resultados["EnvironmentSatisfaction"].set_index("EnvironmentSatisfaction").loc[orden,].index, 
                        height=resultados["EnvironmentSatisfaction"]["EmployeeId"], 
                        color=["#33B5FF", "#FFBB33"], 
                        label="Attrition")

# Título del primer gráfico de barras
axes[0].set_title('Satisfacción con el Ambiente de Trabajo')

# Etiqueta del eje x del primer gráfico de barras
axes[0].set_xlabel('Environment Satisfaction')

# Etiqueta del eje y del primer gráfico de barras
axes[0].set_ylabel('Número de Empleados')

# Graficar el segundo gráfico de barras para JobSatisfaction
axes[1].bar(x=resultados["JobSatisfaction"].set_index("JobSatisfaction").loc[orden, ].index, 
                        height=resultados["JobSatisfaction"]["EmployeeId"], 
                        color=["#33B5FF", "#FFBB33"], 
                        label="Attrition")

# Título del segundo gráfico de barras
axes[1].set_title('Satisfacción con el Trabajo')

# Etiqueta del eje x del segundo gráfico de barras
axes[1].set_xlabel('Job Satisfaction')

# Etiqueta del eje y del segundo gráfico de barras
axes[1].set_ylabel('Número de Empleados');
````

# Gráfico Circular o de Pastel

Es un tipo de visualización que muestra la proporción de partes respecto a un todo. Representado como un círculo dividido en sectores, cada uno representa una categoría y su ángulo es proporcional a la cantidad que representa del total. Útil para mostrar contribuciones relativas y distinguir partes. Es importante tener en cuenta que este tipo de gráfico no lo tenemos en Seaborn

```python
plt.pie(valores, labels=None, autopct=None, startangle=None, data=None, colors=None)
```

- `labels`: Columna con etiquetas de cada sector.

- `valores`: Columna con valores numéricos de cada sector.

- `autopct` (opcional): Muestra los porcentajes en cada sector con formato específico.

- `startangle` (opcional): Controla el ángulo inicial del primer sector.

- `data`: DataFrame de origen de los datos.

- `colors` (opcional): Lista de colores para cada categoría.

- `textprops` (opcional): Diccionario para cambiar el tamaño y tipo de fuente.


📌 **NOTA IMPORTANTE I:** Requiere un procesamiento previo, usualmente un groupby para contar las categorías.          

<div style="background-color: #40B5AD; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>

En esta ocasión, abordaremos nuevamente el tema de la inclusión y diversidad en nuestra empresa. Examinaremos la distribución de género entre los empleados y exploraremos cómo se distribuyen según su nivel educativo. Este análisis nos permitirá comprender mejor la composición de nuestro equipo y tomar medidas para promover la igualdad y la inclusión.</strong>
<div>

````python
# empezamos preparando los datos para poder hacer el pieplot usando un groupby. En este caso tendremos que contar el número de clientes que tenemos por nivel educativo
df_educacion = df.groupby("EducationField")["EmployeeId"].count().reset_index()
display(df_educacion)

df_genero = df.groupby("Gender")["EmployeeId"].count().reset_index()
display(df_genero)
````

````python
# lo primero que vamos a hacer es definir una lista para definir los colores que queremos para cada una de las categorías. 
# Esta lista debe tener tantos elementos como categorías tenemos en la columna de estudio (en este caso 7)
colores_educacion = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal", "turquoise"]
colores_genero = ["darkorange", "g"]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (20, 9))

# creamos el primer gráfico basándonos en el DataFrame que hemos creado en el paso anterior
axes[0].pie("EmployeeId", labels= "EducationField",
        data = df_educacion, 
        autopct=  '%1.1f%%', 
        colors = colores_educacion, 
        textprops={'fontsize': 8}, 
        startangle=90)

# ponemos título a la primera gráfica
axes[0].set_title("Distribución de empleados por Educación");

# creamos el segundo gráfico basándonos en el DataFrame que hemos creado en el paso anterior
axes[1].pie("EmployeeId", labels= "Gender",
        data = df_genero, 
        autopct=  '%1.1f%%', 
        colors = colores_genero, 
        textprops={'fontsize': 10}, 
        startangle=90)

# ponemos título a la segunda gráfica
axes[1].set_title("Distribución de empleados por Género");
````

¿Qué indica el parámetro *autopct*?

- `'%'`: Representa el símbolo de porcentaje (%) en la cadena formateada.

- `'1.1f'`: Describe el formato del número:
  
  - `1`: Ancho mínimo del campo, indicando al menos 1 carácter.
  
  - `.1`: Cantidad de dígitos después del punto decimal a mostrar.
  
  - `f`: Formato de número de punto flotante.
