# ¿Por qué es importante la visualización?

La visualización es importante para los analistas de datos por varias razones:

1. **Comprensión de los datos**: Las visualizaciones permiten explorar y comprender los datos de manera más intuitiva. Al representar los datos gráficamente, es más fácil identificar patrones, tendencias, relaciones y anomalías en los datos que pueden no ser evidentes en una tabla o conjunto de números.

2. **Identificación de insights**: Las visualizaciones pueden revelar insights valiosos ocultos en los datos. Al representar los datos en diferentes formatos gráficos, es posible descubrir patrones, correlaciones o discrepancias que pueden conducir a nuevos conocimientos y oportunidades para el negocio.

3. **Detección de errores y valores atípicos**: Las visualizaciones pueden ayudar a identificar errores en los datos o valores atípicos que pueden afectar los análisis. Al observar los gráficos, es más probable que se detecten discrepancias o valores inusuales que requieran una revisión adicional.

4. **Soporte para la toma de decisiones**: Las visualizaciones brindan una base sólida para la toma de decisiones fundamentada en datos. Al presentar información de manera visual, los analistas pueden respaldar sus argumentos con evidencia visual y facilitar la comprensión y aceptación de sus conclusiones por parte de los tomadores de decisiones.

5. **Comunicación efectiva**: Las visualizaciones ayudan a comunicar los resultados y hallazgos de análisis de datos de manera clara y concisa. Los gráficos y las visualizaciones permiten transmitir información compleja de una manera más accesible y comprensible para las partes interesadas, lo que facilita la toma de decisiones informadas.

# Boxplot y Violinplot

Recordemos que ambos son representaciones gráficas que muestran la distribución de un conjunto de datos numéricos a través de cuartiles. Proporcionan información sobre la mediana, cuartiles y valores atípicos.

La sintaxis de un boxplot en Seaborn es:

```python
sns.boxplot(x, y, data=None, hue=None, width=0.8, palette=None, color=None, order = None, whis=1.5)
```

La sintaxis básica de un violinplot es: 

```python
sns.violinplot(x=None, y=None, data=None, color=None, palette=None, linewidth=1.5)
```



<div style="background-color: #40B5AD; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>
En este análisis, exploraremos si la satisfacción laboral se correlaciona con los salarios de los empleados. A través de la adición de la variable de satisfacción laboral al análisis, examinaremos si aquellos con niveles más altos de satisfacción también tienen salarios más altos. Este enfoque nos permitirá comprender mejor la relación entre la satisfacción laboral y la compensación. </strong>
<div>

````python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 10))

# creamos el código para crear el primer boxplot 
sns.boxplot(x = "JobSatisfaction",
            y = "MonthlyIncome",
            data = df, 
            palette = 'viridis', 
            hue = "JobSatisfaction",
            order = ["Low", "Medium", "High", "Very High"], 
            ax = axes[0])

# ponemos título de la gráfica
axes[0].set_title("Relación entre la variable MonthlyIncome y JobSatisfaction", fontsize = 10)

# cambiamos la etiqueta del eje x
axes[0].set_xlabel("Satisfacción Laboral", fontsize = 10)
axes[0].set_ylabel("Salario Mensual ($)", fontsize = 10)

# movemos la leyenda
axes[0].legend(loc='upper right', # posición de la leyenda
           bbox_to_anchor=(1.1, 1)); # coordenadas exactas de donde colocar la leyenda

# creamos el código para crear el primer boxplot 
sns.violinplot(x = "JobSatisfaction",
            y = "MonthlyIncome",
            data = df, 
            palette = 'viridis', 
            hue = "JobSatisfaction",
            order = ["Low", "Medium", "High", "Very High"], 
            ax = axes[1])

# ponemos título de la gráfica
axes[1].set_title("Relación entre la variable MonthlyIncome y JobSatisfaction", fontsize = 10)

# cambiamos la etiqueta del eje x
axes[1].set_xlabel("Satisfacción Laboral", fontsize = 10)
axes[1].set_ylabel("Salario Mensual ($)", fontsize = 10)

# movemos la leyenda
axes[1].legend().remove(); # eliminamos la leyenda

# ajustamos las graficas
plt.tight_layout()
````

<strong>¿Cómo podemos interpretar esta gráfica?</strong>

En esta nueva visualización, observamos los niveles de satisfacción laboral en relación con los salarios de los empleados. Las principales conclusiones son:

- En ambas gráficas vemos la misma información, añadiendo la información de la densidad de información en el violinplot. 
  
- No se identifican grandes discrepancias entre los niveles de satisfacción y los salarios.
  
- En todos los niveles de satisfacción, hay empleados que reciben remuneraciones notablemente más altas. Estos hallazgos sugieren una necesidad de revisar las políticas de compensación para garantizar la equidad salarial.

# Swarmplot

Es un tipo de gráfico que muestra la distribución de datos de una variable categórica en relación con una variable numérica. Se utiliza para visualizar la distribución de los datos y la densidad de puntos dentro de cada categoría. El Swarmplot es útil para identificar patrones de distribución y densidad de datos en variables categóricas en relación con variables numéricas, siendo especialmente útil en conjuntos de datos con muchos puntos.

La sintaxis básica del Swarmplot en Seaborn es la siguiente:

```python
sns.swarmplot(x, y, data, hue=None, dodge=False, orient=None, color=None, order = None)
```

- `x`: Nombre de la columna con las categorías en el eje x.
  
- `y`: Nombre de la columna con los valores numéricos en el eje y.
  
- `data`: DataFrame o conjunto de datos en uso.
  
- `hue` (opcional): Agrega otra variable categórica para diferenciar los puntos por color.
  
- `dodge` (opcional): Controla la separación de los puntos en categorías distintas si hay más de una variable categórica.
  
- `orient` (opcional): Cambia la orientación del gráfico (horizontal o vertical).
  
- `color` (opcional): Especifica el color de los puntos en el gráfico.

- `order`: Establece el orden en el que queremos que se muestren las categorías. 

<div style="background-color: #40B5AD; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>
En este caso analizaremos el mismo problema que en el ejemplo anterior para comparar los gráficos. Recordemos que ibamos a explorar si la satisfacción laboral se correlaciona con los salarios de los empleados. A través de la adición de la variable de satisfacción laboral al análisis, examinaremos si aquellos con niveles más altos de satisfacción también tienen salarios más altos. Este enfoque nos permitirá comprender mejor la relación entre la satisfacción laboral y la compensación. </strong>
<div>


````python
# establecemos el tamaño de la gráfica
plt.figure(figsize = (15, 5))

# creamos el gráfico
sns.swarmplot(x = "JobSatisfaction",
            y = "MonthlyIncome",
            data = df, 
            palette = 'viridis', 
            hue = "JobSatisfaction",
            order = ["Low", "Medium", "High", "Very High"])

# colocamos la leyenda
plt.legend(loc = "upper right", bbox_to_anchor = (1.12, 1))

# cambiamos el nombre al eje x
plt.xlabel("Satisfacción Laboral")

# cambiamos el nombre de la etiqueta del eje y
plt.ylabel("Salario Mensual ($)");
````

<div style="background-color: #FFBB33; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>¿Cómo podemos interpretar esta gráfica?</strong>

Lo primero que vemos es en general observamos los mismos resultados, pero de una forma diferente. Las principales diferencias entre un violinplot y un swarmplot son: 

- El violinplot utiliza un núcleo de densidad para representar la distribución de los datos. 
  
- El swarmplot coloca cada punto de datos a lo largo de un eje categórico, los puntos no se superponen. Esto permite ver la distribución y la densidad de los datos en cada nivel de forma más precisa. 

Es decir, el violinplot muestra la distribución de los datos de forma suavizada, el swarmplot es una visualización más detallada de la distribución de los datos y sus distribuciones. 
</div>

# Boxenplot en Seaborn

El boxenplot, también conocido como diagrama de caja extendido, es una variante del diagrama de caja que proporciona una representación más detallada de la distribución de los datos. Se utiliza para visualizar la distribución y la concentración de valores en un conjunto de datos, especialmente cuando hay una gran cantidad de datos y se requiere una mayor resolución en las colas de la distribución.

La sintaxis básica es:

```python
sns.boxenplot(x, y, data=None, hue=None, orient=None, color=None, palette = None)
```

- `x` y `y`: Nombres de las variables en los datos que se utilizarán para las etiquetas de los ejes x e y, respectivamente.
  
- `data`: DataFrame o array de datos que contiene las variables de interés.

- `hue` (opcional): Variable categórica en los datos que se utilizará para distinguir las agrupaciones de observaciones.

- `orient` (opcional): Orientación de la gráfica, puede ser "v" (vertical) u "h" (horizontal).

- `color` (opcional): Color de los elementos de la gráfica.

- `palette` (opcional): Paleta de colores que usamos para la gráfica. 

<div style="background-color: #40B5AD; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>
En este caso analizaremos el mismo problema que en el ejemplo anterior para comparar los gráficos. Ibamos a explorar si la satisfacción laboral se correlaciona con los salarios de los empleados. A través de la adición de la variable de satisfacción laboral al análisis, examinaremos si aquellos con niveles más altos de satisfacción también tienen salarios más altos. Este enfoque nos permitirá comprender mejor la relación entre la satisfacción laboral y la compensación.
 </strong>
<div>


````python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 10))

# creamos el código para crear el primer boxplot 
sns.boxplot(x = "JobSatisfaction",
            y = "MonthlyIncome",
            data = df, 
            palette = 'viridis', 
            hue = "JobSatisfaction",
            order = ["Low", "Medium", "High", "Very High"], 
            ax = axes[0])

# ponemos título de la gráfica
axes[0].set_title("Relación entre la variable MonthlyIncome y JobSatisfaction", fontsize = 10)

# cambiamos la etiqueta del eje x
axes[0].set_xlabel("Satisfacción Laboral", fontsize = 10)
axes[0].set_ylabel("Salario Mensual ($)", fontsize = 10)

# movemos la leyenda
axes[0].legend(loc='upper right', # posición de la leyenda
           bbox_to_anchor=(1.1, 1)); # coordenadas exactas de donde colocar la leyenda

# creamos el código para crear el primer boxplot 
sns.boxenplot(x = "JobSatisfaction",
            y = "MonthlyIncome",
            data = df, 
            palette = 'viridis', 
            hue = "JobSatisfaction",
            order = ["Low", "Medium", "High", "Very High"], 
            ax = axes[1])

# ponemos título de la gráfica
axes[1].set_title("Relación entre la variable MonthlyIncome y JobSatisfaction", fontsize = 10)

# cambiamos la etiqueta del eje x
axes[1].set_xlabel("Satisfacción Laboral", fontsize = 10)
axes[1].set_ylabel("Salario Mensual ($)", fontsize = 10)

# movemos la leyenda
axes[1].legend().remove(); # eliminamos la leyenda

# ajustamos las graficas
plt.tight_layout()
````

<div style="background-color: #FFBB33; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>¿Cómo podemos interpretar esta gráfica?</strong>

Lo primero es que todas estas visualizaciones (boxplot, swarmplot, violinplot y boxenplot) son bastante similares entre ellas. Sin embargo, vamos a resumir todas sus diferencias:

| Gráfico      | Descripción                                                                                                                                                               | Uso                                                                                                                                                                                      |
|--------------|---------------------------------------------------------------------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| Violinplot   | Representa la distribución de los datos, mostrando la densidad de probabilidad de los valores.                                                                          | Útil para identificar la forma y simetría de la distribución de los datos.                                                                                                                |
| Swarmplot    | Muestra la distribución de los datos colocando cada punto individual a lo largo del eje categórico.                                                                    | Útil para visualizar la distribución de datos cuando el número de observaciones es relativamente pequeño.                                                                                |
| Boxplot      | Proporciona una representación visual de la distribución de los datos, incluyendo la mediana, los cuartiles y los valores atípicos.                                       | Es eficaz para comparar la distribución de diferentes grupos de datos y para identificar la dispersión y la presencia de valores atípicos.                                               |
| Boxenplot    | Es una variante del boxplot que proporciona una mayor resolución en las colas de la distribución.                                                                       | Útil cuando se trabaja con grandes conjuntos de datos donde se requiere una mayor resolución en las colas de la distribución, y para identificar más valores atípicos en los datos. |

# Stripplot en Seaborn

El stripplot es un tipo de gráfico utilizado para visualizar la distribución de datos unidimensionales, especialmente cuando se comparan diferentes categorías. Muestra puntos individuales a lo largo de un eje categórico, con un ligero desplazamiento aleatorio en el eje categórico para evitar la superposición de puntos. Se utiliza para identificar la distribución y la densidad de los datos, especialmente cuando el número de observaciones es relativamente pequeño.

La sintaxis básica de un stripplot en Seaborn es la siguiente:
```python
sns.stripplot(x, y, data, hue = None, palette = None, alpha = None, marker = "o", size = 5)
```
Donde:
- `x`: Nombre de la columna con las categorías en el eje x.
  
- `y`: Nombre de la columna con los valores numéricos en el eje y.
  
- `data`: DataFrame o conjunto de datos en uso.

- `hue`: Permite agrupar los puntos por otra variable categórica y asignarles diferentes colores en función de esta variable.
  
- `palette`: Define la paleta de colores a utilizar para los puntos cuando se utiliza el parámetro `hue`.

- `alpha`: Define la transparencia de los puntos, permitiendo controlar la opacidad de los mismos.
  
- `marker`: Especifica el tipo de marcador utilizado para representar los puntos en el gráfico.

- `size`: Controla el tamaño de los puntos en el gráfico.

````python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 10))

# creamos el código para crear el primer boxplot 
sns.boxplot(x = "JobSatisfaction",
            y = "MonthlyIncome",
            data = df, 
            palette = 'viridis', 
            hue = "JobSatisfaction",
            order = ["Low", "Medium", "High", "Very High"], 
            ax = axes[0])

# ponemos título de la gráfica
axes[0].set_title("Relación entre la variable MonthlyIncome y JobSatisfaction", fontsize = 10)

# cambiamos la etiqueta del eje x
axes[0].set_xlabel("Satisfacción Laboral", fontsize = 10)
axes[0].set_ylabel("Salario Mensual ($)", fontsize = 10)

# movemos la leyenda
axes[0].legend(loc='upper right', # posición de la leyenda
           bbox_to_anchor=(1.1, 1)); # coordenadas exactas de donde colocar la leyenda

# creamos el código para crear el primer boxplot 
sns.stripplot(x = "JobSatisfaction",
            y = "MonthlyIncome",
            data = df, 
            palette = 'viridis', 
            hue = "JobSatisfaction",
            order = ["Low", "Medium", "High", "Very High"], 
            ax = axes[1])

# ponemos título de la gráfica
axes[1].set_title("Relación entre la variable MonthlyIncome y JobSatisfaction", fontsize = 10)

# cambiamos la etiqueta del eje x
axes[1].set_xlabel("Satisfacción Laboral", fontsize = 10)
axes[1].set_ylabel("Salario Mensual ($)", fontsize = 10)

# movemos la leyenda
axes[1].legend().remove(); # eliminamos la leyenda

# ajustamos las graficas
plt.tight_layout()
````

<div style="background-color: #FFBB33; border-left: 6px solid #000080; padding: 10px; color: black;">

Lo primero es que todas estas visualizaciones (boxplot, swarmplot, violinplot, boxenplot, y stripplot) son bastante similares entre ellas. Sin embargo, vamos a resumir todas sus diferencias: 


| Gráfico      | Descripción                                                                                                                                                               | Uso                                                                                                                                                                                      |
|--------------|---------------------------------------------------------------------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|

| Stripplot    | Muestra la distribución de los datos colocando puntos individuales a lo largo del eje categórico, con cierto desplazamiento aleatorio en el eje categórico para evitar la superposición. | Útil para visualizar la distribución de datos cuando el número de observaciones es relativamente pequeño y cuando se desea mostrar cada punto de datos individualmente.         |

# Pointplot con Seaborn

Es un tipo de gráfico que se utiliza para mostrar la relación entre variables numéricas y categóricas. Se representa la media de los valores numéricos para cada nivel de la variable categórica, junto con un intervalo de confianza o desviación estándar.

La sintaxis básica es:

```python
sns.pointplot(x, y, data=None, hue=None, ci=95, estimator=np.mean, color = None, palette = None)
```

- `x`: Nombre de la columna con las categorías en el eje x.
  
- `y`: Nombre de la columna con los valores numéricos en el eje y.
  
- `data`: DataFrame o conjunto de datos en uso.
  
- `hue` (opcional): Agrega otra variable categórica para diferenciar los puntos por color.
  
- `ci` (opcional): Controla el tamaño del intervalo de confianza alrededor de la media. Por defecto, es 95%.
  
- `estimator` (opcional): Función que se utiliza para estimar la tendencia central de los datos. Por defecto, es la media (`np.mean`). Puede ser cambiado por otras funciones estadísticas como `np.median` o `np.std`.

- `color` (opcional): Para establecer el color de la gráfica. 

- `palette` (opcional): Con este parámetro establecemos la paleta de colores que queremos usar para la visualización. 

<div style="background-color: #40B5AD; border-left: 6px solid #000080; padding: 10px; color: black;">
<strong>
En este análisis, investigaremos si el estado civil y la paternidad afectan los salarios de los empleados. Evaluaremos si existe alguna disparidad salarial basada en el estado civil y la situación parental, lo que nos permitirá abordar posibles inequidades y promover una cultura laboral más inclusiva.
 </strong>
<div>


````python
# generamos el pointplot usando Seaborn
sns.pointplot(x = "Sons", 
              y = "MonthlyIncome", 
              data = df, 
              estimator=np.median, # calculamos la mediana de los salarios de los empleados
              palette = "magma", 
              hue = "MaritalStatus")

# cambiamos las etiquetas del eje x
plt.xticks([0,1], ["Sin hijos", "Con hijos"])

# quitamos el nombre al eje x
plt.xlabel("")

# cambiamos el nombre del eje y
plt.ylabel("Salario Mensual ($)")

# ajustamos la gráfica al espacio que tenemos
plt.tight_layout();
````

En este caso podemos observar una tendencia general de que los empleados con hijos tienden a tener una mediana de salario ligeramente más alta en comparación con aquellos que no tienen hijos en cada estado civil. Específicamente:
