# ¿Por qué es importante la visualización?

La visualización es importante para los analistas de datos por varias razones:

1. **Comprensión de los datos**: Las visualizaciones permiten explorar y comprender los datos de manera más intuitiva. Al representar los datos gráficamente, es más fácil identificar patrones, tendencias, relaciones y anomalías en los datos que pueden no ser evidentes en una tabla o conjunto de números.

2. **Identificación de insights**: Las visualizaciones pueden revelar insights valiosos ocultos en los datos. Al representar los datos en diferentes formatos gráficos, es posible descubrir patrones, correlaciones o discrepancias que pueden conducir a nuevos conocimientos y oportunidades para el negocio.

3. **Detección de errores y valores atípicos**: Las visualizaciones pueden ayudar a identificar errores en los datos o valores atípicos que pueden afectar los análisis. Al observar los gráficos, es más probable que se detecten discrepancias o valores inusuales que requieran una revisión adicional.

4. **Soporte para la toma de decisiones**: Las visualizaciones brindan una base sólida para la toma de decisiones fundamentada en datos. Al presentar información de manera visual, los analistas pueden respaldar sus argumentos con evidencia visual y facilitar la comprensión y aceptación de sus conclusiones por parte de los tomadores de decisiones.

5. **Comunicación efectiva**: Las visualizaciones ayudan a comunicar los resultados y hallazgos de análisis de datos de manera clara y concisa. Los gráficos y las visualizaciones permiten transmitir información compleja de una manera más accesible y comprensible para las partes interesadas, lo que facilita la toma de decisiones informadas.

# Histograma

## Histograma con Seaborn

El histograma, o *histplot* en Seaborn, es una representación gráfica de la distribución de una variable numérica. Divide el rango de valores de la variable en intervalos llamados "bins" y muestra la frecuencia o conteo de observaciones que caen en cada `bin`. Esto permite visualizar la forma y la concentración de los datos en la variable, identificar patrones y detectar posibles valores atípicos.

- El eje x del histograma representa los valores de la variable numérica.
  
- El eje y muestra la frecuencia o densidad de los valores en cada `bin`.

Utilizamos un histograma cuando deseamos representar la distribución de una variable numérica. Es crucial entender cómo se distribuyen los valores y qué patrones o tendencias pueden estar presentes en los datos.

Aquí está cómo hacer esta visualización con Seaborn:

```python
sns.histplot(data, x="variable", bins=10, kde=False, hue = None, color="blue")
```

Donde:

- `data`: El DataFrame o la Serie que contiene los datos.
  
- `x`: La variable numérica (columna) que se desea representar en el eje x del histograma.
  
- `bins` (opcional): El número de intervalos (bins) en los que se dividirá el rango de valores. Puede ser un entero o una lista de valores para especificar los límites de los bins.
  
- `kde` (opcional): Un valor booleano que indica si se desea trazar una estimación de la densidad kernel junto con el histograma (por defecto es `False`).

- `hue`:Diferencia datos basándose en una variable categórica adicional, creando sub-histogramas para cada categoría correspondiente.
  
- `color` (opcional): El color que se utilizará para el histograma (por defecto es `"blue"`).

## Histograma con matplotlib


Hemos hecho el histograma con Seaborn. Hagámoslo ahora con matplotlib, para poder comparar las diferencias entre librerías. Su sintaxis básica es:

```python
plt.hist(x, bins=None, range=None, density=False, cumulative=False, color=None, edgecolor=None)
```

Los parámetros principales son:

- `x`: Los datos numéricos que se usarán para crear el histograma.

- `data`: El DataFrame o la Serie que contiene los datos.

- `bins`: Especifica el número de contenedores (bins) en los que se dividirán los datos. También se puede proporcionar una secuencia de valores que representan los bordes de los bins.

- `range`: Define el rango de valores que se considerarán en el histograma. Se pasa como una lista, indicando el valor mínimo y máximo en el eje x.

- `density`: Si es True, el histograma mostrará densidades en lugar de frecuencias.

- `cumulative`: Si es True, el histograma mostrará una distribución acumulativa en lugar de una distribución normal.

- `color`: Color de las barras del histograma. Se pueden utilizar los mismos colores que en Seaborn.

- `edgecolor`: Color del borde de las barras del histograma.

Además de estos argumentos, hay otros opcionales que se pueden utilizar para personalizar el histograma, como el grosor de las barras, la transparencia, la orientación, entre otros.


````python
# empecemos construyendo una gráfica básica con matplotlib
# a priori, la gráfica es igual, solo que con una apariencia un poco diferente. 
# La principal diferencia es que no tenemos los nombres de los ejes que si teníamos en la gráfica de seaborn (Age y count)
plt.hist(x = "Age", 
         data = df);
````

````python
# modifiquemos un poco la gráfica con algunos de sus parámetros
# en este caso hemos añadido los 'bins', el 'color' y el 'edgecolor', de forma similar a como lo hicimos en seaborn
plt.hist(x = "Age", 
         data = df, 
         bins = 20, 
         color = "violet", 
         edgecolor = "black", # poner color al borde de las barras
         range = [15, 70], # cambiar el límite del eje X 
         density = True); # mostramos densidades en vez de frecuencias 
````

***Diferencia entre Frecuencia y Densidad***

En un histograma, la frecuencia y la densidad representan la distribución de datos, pero difieren en la escala del eje y y el significado asociado a cada una:

- **Frecuencia**: Indica el recuento de observaciones en cada bin o intervalo. La altura de las barras representa el número de observaciones en cada bin.

- **Densidad**: Muestra la proporción relativa de observaciones en cada bin en relación con el tamaño total de los datos. La densidad muestra la probabilidad relativa de que una observación pertenezca a un bin específico.

Como hemos dicho, la principal diferencia entre Seaborn y Matplotlib en esta gráfica es que en esta última no vemos nos nombres de los ejes. Para añadir etiquetas a los ejes de las gráficas en matplotlib, empleamos `plt.xlabel()` y `plt.ylabel()`.

1. `plt.xlabel('Etiqueta del eje x')`: Agrega una etiqueta al eje x del gráfico. Se especifica el texto que se mostrará como etiqueta.

    ```python
    plt.xlabel('Etiqueta del eje x', # 'Etiqueta del eje x' es el texto deseado para la etiqueta del eje x.
               fontsize = 8 ) # el tamaño de la fuente  
    ```

2. `plt.ylabel('Etiqueta del eje y')`: Añade una etiqueta al eje y del gráfico. Permite especificar el texto de la etiqueta.

    ```python
    plt.ylabel('Etiqueta del eje y', # 'Etiqueta del eje y' es el texto deseado para la etiqueta del eje y.
                fontsize = 8 )  # el tamaño de la fuente
    ```


````python
# modifiquemos un poco la gráfica con algunos de sus parámetros
# en este caso hemos añadido los 'bins', el 'color' y el 'edgecolor', de forma similar a como lo hicimos en seaborn
plt.hist(x = "Age", 
         data = df, 
         bins = 20, 
         color = "violet", 
         edgecolor = "black", 
         range = [15, 70], 
         density = True); # mostramos densidades en vez de frecuencias 

# añadimos el método 'plt.xlabel()' para ponerle nombre al eje x
plt.xlabel("Edad")

# añadimos el método 'plt.ylabel()' para ponerlo nombre al eje y
plt.ylabel("Densidad");
````

📌 **NOTA**: Seaborn y Matplotlib comparten algunos métodos, especialmente los relacionados con la modificación del estilo de las gráficas. En este sentido, los métodos `plt.xlabel()` y `plt.ylabel()` que acabamos de ver también se pueden utilizar con gráficas de Seaborn de la misma manera.

````python
# creamos la gráfica de seaborn
sns.histplot(x = "Age", 
             data = df, 
             color = "violet", 
             hue = "Attrition", 
             palette= "cubehelix",
             kde = True, 
             bins = 30 )

# usando 'plt.xlabel()' cambiamos el nombre del eje x
plt.xlabel("Edad")

# usando el método 'plt.ylabel()' cambiamos el nombre del eje y
plt.ylabel("Número de Empleados");
````

# Gráfico de dispersión o *scatterplot*

## Scatterplot con Seaborn

Un gráfico de dispersión, también llamado diagrama de dispersión o gráfico XY, muestra la relación entre dos variables. Cada punto representa un par de valores de estas variables. Su posición indica los valores correspondientes. Este tipo de gráfico ayuda a identificar patrones, tendencias o correlaciones entre las variables. Puede responder preguntas como: ¿hay una relación lineal? ¿Es positiva o negativa? ¿Hay valores atípicos?

En Seaborn, su sintaxis básica es:

```python
sns.scatterplot(x, y, data=None, hue=None, style=None, size=None, markers=True, palette=None, color = None)
```

Parámetros principales:

- `x` e `y`: Variables numéricas para los ejes x e y.

- `data`: El DataFrame con los datos.

- `hue` (opcional): Agrega una dimensión a través de diferentes colores para categorías.

- `style`: Por defecto es None, lo que significa que no se utiliza ninguna variable categórica para distinguir los estilos visuales de los puntos.

- `markers` (opcional): Cambia la forma de los puntos. Algunos de los markers que podemos usar son: 

    - `'.'` Punto
    - `','` Píxel
    - `'o'` Círculo (por defecto)
    - `'v'` Triángulo hacia abajo
    - `'^'` Triángulo hacia arriba
    - `'<'` Triángulo hacia la izquierda
    - `'>'` Triángulo hacia la derecha
    - `'1'` Triángulo hacia abajo delgado
    - `'2'` Triángulo hacia arriba delgado
    - `'3'` Triángulo hacia la izquierda delgado
    - `'4'` Triángulo hacia la derecha delgado
    - `'s'` Cuadrado
    - `'p'` Pentágono
    - `'*'` Estrella
    - `'+'` Cruz
    - `'x'` Cruz diagonal
    - `'D'` Diamante
    - `'d'` Diamante delgado

# Gráfico de Regresión o *regplot*

El gráfico de regresión, o *regplot*, combina un diagrama de dispersión (*scatter plot*) con una línea de regresión lineal ajustada a los datos. Esta línea de regresión muestra la tendencia general de la relación entre dos variables, ayudando a identificar patrones o tendencias.

Usamos este tipo de gráfico para **explorar la relación entre dos variables numéricas y ver cómo se ajusta una línea de regresión lineal a esos datos**. 

Su sintaxis básica es:

```python
sns.regplot(x, y, data=None, marker='o', color = None)
```

- `x`: Especifica la columna que se colocará en el eje x (horizontal).
  
- `y`: Indica la columna que se colocará en el eje y (vertical).
  
- `data`: Es el DataFrame que contiene los datos.
  
- `markers` (opcional): Permite cambiar los puntos por otro tipo de marcador en la gráfica, similar al *scatter plot*.
  
- `color` (opcional): Permite cambiar el color de los puntos de la gráfica.

Para ver las diferencias con el *scatter plot*, usaremos la misma pregunta que en el ejemplo anterior:

</div>


📌 **NOTA:** Para este tipo de gráfico, no contamos con un método equivalente en `matplotlib`.

````python
# creamos una gráfica usando el 'sns.regplot()' para evaluar la relación que existe entre dos variables numéricas
sns.regplot(x = "TotalWorkingYears", 
            y = "MonthlyIncome", 
            data = df, 
            marker = "d", 
            color = "purple", 
            scatter_kws={'s':2})

# cambiamos los nombres de los ejes como hemos estado haciendo hasta ahora
plt.xlabel("Años de experiencia", fontsize = 8)
plt.ylabel("Salario Mensual", fontsize = 8)

# ponemos título a la gráfica
plt.title("Relación entre los Años de Experiencia y el Salario")

# quitamos la linea de arriba y de la derecha
plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;
````

# Gráfico de cajas (boxplot) y violín

Ambos son representaciones gráficas que muestran la distribución de un conjunto de datos numéricos a través de cuartiles. Proporcionan información sobre la mediana, cuartiles y valores atípicos.

Antes de continuar, definamos los *outliers* o valores atípicos. Los *outliers* son observaciones que están muy alejadas del resto de los datos en un conjunto o muestra estadística, indicando anormalidades o extremos en los datos.

Es crucial identificar los *outliers* por varias razones:

1. **Impacto en el análisis estadístico**: Pueden afectar los resultados de los análisis, distorsionando estadísticas como la media y la desviación estándar.

2. **Detección de errores o problemas**: Indican errores en la recopilación o entrada de datos, así como problemas en el proceso de medición.

3. **Identificación de casos especiales**: Algunos *outliers* pueden representar casos únicos o excepcionales, proporcionando información valiosa.

## Boxplot con Seaborn

Un boxplot es una representación gráfica que resume la distribución y tendencia central de los datos numéricos, además de identificar *outliers*.

La sintaxis de un boxplot en Seaborn es:

```python
sns.boxplot(x, y, data=None, hue=None, width=0.8, palette=None, color=None, order = None, whis=1.5)

```
Donde: 
- `x`: Nombre de la columna que queremos poner en el eje x (horizontal). Por lo general, esta columna contiene categorías o grupos que deseas comparar. Esto solo lo usaremos si queremos establecer la relación entre una variable numérica y una categórica, veremos un ejemplo más adelante en la lección. 

- `y`: Nombre de la columna que tiene los valores numéricos que queremos representar en el eje y (vertical). Estos valores se utilizarán para construir los boxplots y mostrar cómo se distribuyen los datos.

- `data`: DataFrame que contiene tus datos. 

- `hue`: Permite agregar una dimensión adicional de agrupación y colorea los boxplots según la categoría especificada. Esto solo lo usaremos cuando queramos establecer la relación entre variable categórica y numérica, que como hemos dicho lo veremos más adelante. 

- `width`: Define el ancho relativo de los boxplots.

- `color`: El color que queremos en el violinplot. Este parámetro lo usaremos cuando estemos analizando solo UNA variable numérica. 

- `palette`: La paleta de colores que queremos en nuestra gráfica. Usaremos este parámetro cuando estemos analizando la relación entre una variable numérica y una categórica. Veremos un ejemplo un poco más adelante en la lección.

- `order`: Controla el orden de las categorías en el eje x.

- `whis`: Define el tamaño del bigote en relación con el rango intercuartílico (IQR). Por defecto es `1.5`, lo que significa que los bigotes se extienden hasta 1.5 veces el IQR desde el primer y tercer cuartil. Este parámetro puede ajustarse para cambiar la sensibilidad a los valores atípicos en el gráfico del diagrama de caja.

**Características de los boxplots:**

- La caja central representa el rango intercuartílico (IQR), desde el primer cuartil (Q1) hasta el tercero (Q3).
  
- La línea en la caja es la mediana (Q2).
  
- Los "bigotes" extienden desde la caja hasta los valores más extremos dentro de un rango determinado.
  
- Valores fuera de los bigotes se consideran outliers.

Para hacer esto usaremos el método `plt.subplots()`, el cual se utiliza para crear una figura y una o varias gráficas (subplots). Por lo tanto, este método es útil cuando se desea crear múltiples gráficos en una misma figura.

Su sintaxis básica es:

```python
fig, axes = plt.subplots(nrows, ncols, sharex=False, sharey=False, figsize=None)
```

- `nrows`: Número de filas de subplots en la figura.

- `ncols`: Número de columnas de subplots en la figura.

- `sharex` (opcional): Si es `True`, los subplots compartirán el mismo eje x.

- `sharey` (opcional): Si es `True`, los subplots compartirán el mismo eje y.

- `figsize` (opcional): Tamaño de la figura en pulgadas (ancho, alto).

Para ver un ejemplo de los subplots juntaremos las dos gráficas que hemos creado anteriormente. 

````python
# establecemos el número de filas, columnas que queremos así como el tamaño que vamos a querer que tenga la figura. 
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 5))

# al ejecutar esta primera línea vemos que tenemos una figura con dos gráficas en su interior. 
````

````python
# establecemos el número de filas, columnas que queremos así como el tamaño que vamos a querer que tenga la figura. 
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 5))


# Vamos a empezar a insertar alguna gráfica, en concreto vamos a rellenar la primera (la de la izquierda) con el boxplot que creamos anteriormente
# fijaos como no cambia nada, lo único que hemos hecho ha sido añadir el parámtro 'ax' para hacer referencia en cual de los subplots lo queremos
# en este caso pondremos ax = 0, porque ya sabemos que en Python todo empieza siempre en 0. 
sns.boxplot(x = "MonthlyIncome", 
            data = df, 
            color ="violet", 
            ax = axes[0]); # especificamos en cual de las cajas queremos "insertar" esta gráfica
````

````python
# YA HEMOS AÑADIDO UNA GRÁFICA EN NUESTRO SUBPLOT, PERO NO TIENE NINGUNA DE LAS MODIFICACIONES QUE HICIMOS EN LA ORIGINAL.
# establecemos el número de filas, columnas que queremos así como el tamaño que vamos a querer que tenga la figura. 
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 5))

# añadimos el gráfico
sns.boxplot(x = "MonthlyIncome", 
            data = df, 
            color ="violet", 
            ax = axes[0])

# para añadir un título usaremos el parámetro '.set_title()' 
axes[0].set_title("Distribución de la variable MonthlyIncome")

# cambiamos el nombre del eje x e y usando el método '.set_xlabel()' y '.set_ylabel()' respectivamente. 
axes[0].set_xlabel("Salario Mensual ($)")

# quitamos la linea de la derecha de la gráfica
axes[0].spines['right'].set_visible(False)

# quitamos la linea de arriba de la gráfica
axes[0].spines['top'].set_visible(False)

# quitamos la linea de la izquierda de la gráfica
axes[0].spines['left'].set_visible(False);
````

Ya tenemos una de las gráficas de nuestro subplot, vamos a añadir la segunda, que recordemos es el *boxplot*, como podréis imaginar lo único que tendremos que hacer es añadir el parámetro 'ax' en la gráfica que creamos previamente.

# Lineplot

Es especialmente útil para mostrar tendencias o cambios en el tiempo. Por ejemplo, podemos usar un lineplot para visualizar la evolución de la temperatura a lo largo de diferentes meses del año.

La sintaxis básica para crear un lineplot con Seaborn es:

```python
sns.lineplot(x=None, y=None, data=None, hue=None, style=None, size=None, markers=True, estimator='mean', ci=95, err_style='band')

```

Los principales parámetros que podemos usar son:

- `data`: El DataFrame o conjunto de datos que contiene los datos a visualizar.
  
- `x`: La variable en el eje x.
  
- `y`: La variable en el eje y.
  
- `hue` (opcional): Una variable categórica que será usada para distinguir diferentes categorías en el gráfico. Por ejemplo, si estás visualizando la temperatura a lo largo del tiempo, el parámetro `hue` podría ser utilizado para distinguir entre diferentes ubicaciones geográficas.

- `style` (opcional): Una variable categórica que determina diferentes estilos visuales para cada categoría.
  
- `size` (opcional): Una variable numérica que determina el tamaño de los puntos o líneas en el gráfico.
  
- `markers` (opcional): Una lista de marcadores o un diccionario que asigna marcadores a las categorías.
  
- `estimator` (opcional): Es el estadístico que se va a usar para mostar los datos. Por defecto, media. Es decir, se calculará la media para cada agrupación. Por ejemplo, si ponemos np.median se calculará la mediana para cada grupo. 
  
- `err_style` (opcional): El estilo de las barras de error.

- `errorbar` (opcional): Determina cómo se muestran las barras de error. Puede tomar los siguientes valores:
  
  - `'ci'` (por defecto): Muestra el intervalo de confianza alrededor de la media.
  
  - `'sd'`: Muestra la desviación estándar de los datos.

  - `None`: No muestra barras de error.

# Pairplot

El `pairplot` crea una matriz de gráficos de dispersión para todas las combinaciones de variables en un DataFrame. Es útil para explorar la relación entre múltiples variables en un conjunto de datos de manera rápida y eficiente.

La sintaxis básica es la siguiente:

```python
sns.pairplot(data, hue=None, palette=None, kind='scatter', diag_kind='auto', height=2.5)
```

- `data`: DataFrame que contiene las variables a visualizar.
  
- `hue` (opcional): Variable categórica para diferenciar las visualizaciones por color.
  
- `palette` (opcional): Paleta de colores a utilizar para diferenciar las categorías en `hue`.
  
- `kind` (opcional): Tipo de gráfico para las visualizaciones de los pares (por defecto, `'scatter'`).
  
- `diag_kind` (opcional): Tipo de gráfico para las visualizaciones en la diagonal (por defecto, `'auto'`).
  
- `height` (opcional): Altura de cada gráfico en pulgadas.

**Ventajas e Inconvenientes:**

**Ventajas:**

- Proporciona una vista rápida y general de la relación entre múltiples variables en un conjunto de datos.
  
- Es fácil de usar y proporciona muchas opciones de personalización.
  
- Permite identificar patrones y correlaciones entre variables de forma visual.

**Inconvenientes:**

- Puede ser computacionalmente costoso con grandes conjuntos de datos.
  
- La visualización puede ser abrumadora con un gran número de variables.
  
- No proporciona información sobre la fuerza o dirección de las relaciones entre variables, solo visualiza la presencia de relaciones.


````python
# creamos nuestro gráfico de pairplot
sns.pairplot(data = df[["MonthlyIncome", "DistanceFromHome", "YearsSinceLastPromotion", "TotalWorkingYears", "Attrition"]], 
             hue = "Attrition");
plt.tight_layout();
````
