# Codificación de variables categóricas


El *encoding* es una parte fundamental del preprocesamiento de datos en *machine learning*. Antes de entrenar un modelo, es necesario preparar y transformar los datos de una forma adecuada para su uso.

El **encoding** o codificación se refiere al proceso de convertir datos categóricos (variables que representan categorías) en una forma numérica que pueda ser utilizada por algoritmos de *machine learning*. Los datos categóricos son aquellos que representan características que no son numéricas, como colores, tipos de productos, géneros, etc.

## Importancia del *encoding* en *Machine Learning*

1. **Requisito de algoritmos**: La mayoría de los algoritmos de *machine learning* requieren que los datos de entrada sean numéricos. Esto significa que necesitaremos convertir cualquier tipo de datos no numérico en una forma que los algoritmos puedan entender y procesar.

2. **Conservación de información**: Es importante que el proceso de *encoding* conserve la información contenida en las categorías originales. Esto significa que dos categorías distintas deben ser representadas por valores numéricos distintos.

3. **Interpretación de los Resultados**: La codificación adecuada también facilita la interpretación de los resultados del modelo. Cuando los datos están en forma numérica, es más fácil comprender cómo las diferentes características afectan a la salida del modelo.


Algunos de los métodos más comunes que tenemos para realizar la codificación de las variables categóricas son: 


| Método de Encoding   | Descripción                                                                                   | Cuándo usarlo                                                                      |
|----------------------|-----------------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------|
| One-Hot Encoding     | Cada categoría se convierte en una columna binaria.                                           | - Variables categóricas con pocas categorías.<br>- Cuando no hay una relación ordinal entre las categorías.                                      |
| Label Encoding       | Cada categoría se asigna a un número entero único.                                            | - Variables categóricas con un orden natural.<br>- Algoritmos que pueden interpretar relaciones ordinales entre categorías.         |
| Target Encoding      | Cada categoría se codifica según alguna estadística (p. ej., la media) de la variable objetivo. | - Cuando se quiere mantener el número de características bajo control.<br>- Variables categóricas con muchas categorías.                                     |
| get_dummies          | Método de pandas para generar columnas dummy de variables categóricas.                        | - Cuando se trabaja con pandas y se necesita realizar One-Hot Encoding.             |
| Ordinal Encoding     | Cada categoría se codifica como un número entero, respetando un orden específico.            | - Cuando las categorías tienen un orden inherente.<br>- Variables categóricas con un orden definido.                                     |
| Frequency Encoding   | Las categorías se codifican según su frecuencia de aparición.                                 | - Cuando la frecuencia de una categoría puede estar relacionada con la variable objetivo. |



# Establecer el tipo de variables

Antes de ponernos manos a la obra lo primero que vamos a hacer es recordar qué tipos de variables categóricas nos podemos encontrar. 

- **Variables ordinales**: En las variables ordinales existe un orden entre sus categorías. Por ejemplo, "Nivel educativo" podría tener un orden de "Escolaridad básica", "Educación secundaria", "Educación superior" y "Doctorado".

- **Variables nominales**: Las variables nominales no tienen un orden entre sus categorías. Por ejemplo, "Colores" o "Tipos de Plantas" probablemente no tengan un orden natural entre sus categorías.


Una vez recordado esto, lo siguiente que nos podemos preguntar es ¿cómo podremos saber si una variable tiene orden o no? Para determinar si una variable categórica tiene un orden intrínseco o no, podemos realizar algunas comprobaciones y análisis, aunque no podemos decir que esto sea una receta universal:

1. **Observación de los valores únicos:** Primero, podemos observar los valores únicos presentes en la variable categórica. Si los valores representan categorías que tienen un orden natural, es probable que la variable tenga un orden inherente. Por ejemplo, si la variable representa tamaños como "pequeño", "mediano" y "grande", estos valores tienen un orden natural y  probablemente tenga orden en nuestro conjunto de datos, aunque no necesariamente tiene que ser siempre así. 

2. **Contexto de los datos:** Es importante considerar el contexto de nuestros datos para determinar si existe un orden lógico para las categorías. Algunas variables categóricas pueden tener un orden implícito en función del contexto. Por ejemplo, en una variable que representa el nivel educativo ("primaria", "secundaria", "universitaria"), existe un orden lógico. Sin embargo, igual que el caso anterior, no necesariamente tiene que ser así, y siempre es recomendable explorar los dos siguientes puntos para confirmar o rechazar el orden de las categorías. 

3. **Consulta con expertos:** Si no estamos seguros de si una variable categórica tiene un orden implícito, podemos consultar con expertos en el problema. Ellos pueden proporcionar información valiosa sobre la naturaleza de los datos y si existe un orden subyacente en las categorías.

3. **Análisis Visual:** La visualización de los datos puede ser útil para identificar un posible orden en las categorías. Los gráficos como los diagramas de barras ordenados o los diagramas de caja pueden ayudar a visualizar la distribución de los valores y cualquier tendencia de orden.


4. **Análisis estadístico:** También podemos realizar un análisis estadístico de la variable categórica para identificar cualquier patrón de orden. Por ejemplo, podemos calcular medidas de tendencia central como la media o la mediana para cada categoría y observar si existe un orden coherente en estas medidas.

1. **Visualización de la distribución:** El boxplot proporciona una representación visual clara de la distribución de los valores numéricos dentro de cada categoría. ## Análisis estadístico

Las pruebas estadísticas nos pueden ayudar a determinar si las diferencias que hemos identificado visualmente en el paso anterior son estadísticamente significativas o no, lo que puede ser crucial por varios motivos:

- Nos permite comprender mejor la estructura de los datos y las posibles relaciones entre las variables.


- Puede mejorar la precisión de los modelos predictivos al considerar el orden en variables categóricas como predictores.

<div style="background-color: #F74646; padding: 10px; border-left: 6px solid #000080; color: black; border-radius: 10px;">
<b>Para evaluar si una variable categórica tiene un orden o no, podemos considerar usar las siguientes pruebas estadísticas:</br>
<p></p>

| Prueba                   | Descripción                                                                                                     | Cuándo usarla                                                                                                 | Requisitos estadísticos                                                                                           |
|--------------------------|-----------------------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------------------|
| Prueba de Jonckheere-Terpstra | Esta prueba no paramétrica se utiliza para evaluar si hay una tendencia ordenada significativa entre más de dos grupos. | Se usa cuando queremos determinar si hay un orden significativo entre tres o más grupos ordenados.                  | No requiere asunciones específicas.                                                                              |
| ANOVA (Análisis de Varianza) | Esta prueba paramétrica se utiliza para comparar las medias de tres o más grupos para determinar si son iguales o no. | Es apropiada cuando queremos evaluar si hay diferencias significativas entre las medias de múltiples grupos.     | Requiere que los datos sigan una distribución normal y que las varianzas entre los grupos sean homogéneas.        |
| Prueba de Chi-cuadrado | Esta prueba se utiliza para determinar si hay una asociación significativa entre dos variables categóricas.        | Se usa cuando queremos evaluar si existe una relación significativa entre dos variables categóricas.             | Requiere que las observaciones sean independientes y que las frecuencias esperadas sean mayores que cinco.       |
| Prueba de Wilcoxon | Esta prueba no paramétrica se utiliza para comparar dos muestras relacionadas y determinar si hay una diferencia significativa entre ellas. | Se utiliza cuando queremos comparar dos muestras relacionadas y las condiciones de normalidad no se cumplen.   | No requiere asunciones específicas.                                                                              |
| Prueba de Mann-Whitney | Similar a la prueba de Wilcoxon, esta prueba no paramétrica se utiliza para comparar dos muestras independientes. | Es apropiada cuando queremos comparar dos muestras independientes y las condiciones de normalidad no se cumplen. | No requiere asunciones específicas.                                                                              |
| Prueba de Kruskal-Wallis | Esta prueba no paramétrica es una extensión del ANOVA y se utiliza para comparar tres o más grupos independientes. | Se usa cuando queremos comparar tres o más grupos independientes y las condiciones de normalidad no se cumplen. | No requiere asunciones específicas.                                                                              |
</div>


<div style="background-color: #F74646; padding: 10px; border-left: 6px solid #000080; color: black; border-radius: 10px;">

**NOTA IMPORTANTE ANTES DE CONTINUAR**: 

En este caso, no podremos chequear la independencia de las variables (entre la variable respuesta y la variable predictora categórica). Para asegurar la independencia de las observaciones debemos tener un buen diseño experimental. Para conseguir esto, se deben seguir prácticas como el muestreo aleatorio y la asignación aleatoria de tratamientos a los grupos experimentales. Estas prácticas nos garantizarán que las observaciones en el conjunto de datos sean independientes entre sí y no estén influenciadas por factores externos no controlados.

Llegados a este punto, cuando aprendimos la herramienta de A/B Testing vimos estrategias para evaluarla (el chi-cuadrado y la correlación de Pearson). Recordemos que, el chi-cuadrado es para comparar dos variables categóricas y que la correlación de Pearson es para comparar dos variables numéricas. Nosotros tenemos una variable categórica y una numérica. Es por esto, que en este caso no podremos evaluar la independencia entre las observaciones. 
</div>

```python
# instanciamos la clase de Asunciones, para poder decir qué método usar para el análisis estadístico
asunciones = se.Asunciones(dataframe = df, columna_numerica = "PageValues")
print("\n-------------\n")


for categoria in lista_col_categ:   

    print(f"Estamos analizando la variable {categoria.upper()}")
    
    # comprobamos la asunción de normalidad
    asunciones.identificar_normalidad(metodo = "kolmogorov") 

    # comprobamos la homogeneidad de varianzas
    asunciones.identificar_homogeneidad(columna_categorica = categoria)

    # instanciamos la clase para evaluar si hay diferencias entre los distintos grupos de las variables categóricas
    test = se.TestEstadisticos(df, "PageValues", categoria)
    test.run_all_tests()
    print("\n###########################\n")
```

<div style="background-color: #F74646; padding: 10px; border-left: 6px solid #000080; color: black; border-radius: 10px;">

**¿Por qué hemos elegido el test de Kruskal?**: 
- En nuestro caso, la variable respuesta ("Page Values") no es normal

- Entre las categorías de las variables categóricas no hay homogeneidad de varianzas. 

Por lo tanto, tendremos que usar el test de Wilcoxon, el de MannWithney o el de Kruskal. Sin embargo, el test de Wilcoxon NO admite muestras con distintos tamaños (por lo que no podremos usarlo) y el test de MannWithney no acepta más de dos grupos para comprar. Por todo esto, solo nos queda un test que podemos usar, el test de Kruskal. 
</div>

<div style="background-color: #FFBB33; padding: 10px; border-left: 6px solid #000080; color: black; border-radius: 10px;">
<strong>Interpretación de los resultados de los test estadísticos</strong>

Para entender los resultados de los test estadísticos:

- **Estadístico de prueba**: Es el valor del estadístico de prueba obtenido del test. Este valor se utiliza para evaluar si hay diferencias significativas entre los grupos.
  
- **Valor p**: Es el valor p asociado al estadístico de prueba. Indica la probabilidad de obtener el estadístico de prueba observado, suponiendo que no hay diferencias reales entre los grupos. Un valor p menor que el nivel de significancia (generalmente 0.05) sugiere que hay diferencias significativas entre, al menos, dos de los grupos comparados. 

</div>




<div style="background-color: #FFBB33; padding: 10px; border-left: 6px solid #000080; color: black; border-radius: 10px;">
<strong>Interpretación de los resultados del Test de Tukey</strong>

Para entender los resultados de la tabla que nos devuelve el test:

- **meandiff**: Diferencia media entre los grupos. Indica cuánto difiere la media de la variable respuesta entre los grupos comparados.

- **p-adj**: Valor de p ajustado (p-value adjusted). Es el valor p corregido para múltiples comparaciones. Un valor p ajustado menor que el umbral alfa (generalmente 0.05) indica que la diferencia entre los grupos es estadísticamente significativa.

- **lower**: Límite inferior del intervalo de confianza para la diferencia de medias.

- **upper**: Límite superior del intervalo de confianza para la diferencia de medias.

- **group_diff**: Describe las diferencias entre los grupos comparados. Por ejemplo, para la variable "VISITORTYPE", "New_Visitor-Other" indica la diferencia entre el grupo de nuevos visitantes y el grupo "Other".
