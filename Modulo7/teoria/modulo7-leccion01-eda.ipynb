# Preprocesamiento y Data Cleaning 

El preprocesamiento de datos es fundamental en la ciencia de datos, especialmente para problemas de *machine learning*. Implica preparar y limpiar el conjunto de datos para que sea más adecuado para los algoritmos. Este proceso puede reducir la complejidad, prevenir el sobreajuste y mejorar el rendimiento general del modelo. Es importante destacar que la fase de preprocesamiento se trata de el paso más largo e importante en estos problemas, que aproximadamente ocupa el 70% del tiempo. Este paso es importante porque en la mayoría de las situaciones, los datos proporcionados por el cliente tienen una mala calidad o simplemente no pueden ser alimentados directamente a algún tipo de modelo de ML. Una frase muy usada en este contexto y que veréis muy a menudo, dice: "*Garbage in, garbage out*" (GIGO, por sus siglas en inglés). En otras palabras, si alimentas tu modelo con datos miserables, no esperes que funcione bien.

La fase de preprocesamiento de datos comienza con la comprensión de las sutilezas de su conjunto de datos y los principales problemas de los datos a través del Análisis Exploratorio de Datos. Los datos del mundo real a menudo presentan inconsistencias, errores tipográficos, datos faltantes y diferentes escalas. Debemos abordar estos problemas para hacer que los datos sean más útiles y comprensibles. Este proceso de limpieza y resolución de la mayoría de los problemas en los datos es lo que llamamos el paso de preprocesamiento de datos.

Las técnicas de preprocesamiento de datos incluyen limpieza de datos, reducción de dimensionalidad, ingeniería de características, muestreo de datos, transformación y manejo de datos desequilibrados. 


## Técnicas Comunes de Preprocesamiento de Datos

- **Manejo de Valores Nulos**: Los datos faltantes en el conjunto de datos pueden llevar a resultados engañosos. Por lo tanto, es esencial manejar adecuadamente los valores faltantes. Las técnicas para manejar valores faltantes incluyen la eliminación, eliminando las filas con valores faltantes, y la imputación, reemplazando los valores faltantes con medidas estadísticas como la media, la mediana o el modelo. Este paso es crucial para garantizar la calidad de los datos utilizados para entrenar modelos de aprendizaje automático.

- **Identificación y Manejo de *outliers***: La gestión de outliers en machine learning es esencial para evitar la distorsión de resultados y garantizar la efectividad del modelo. Los outliers, valores extremos en los datos, pueden sesgar el modelo y provocar predicciones inexactas. Al eliminar o transformar estos valores atípicos, se mejora la precisión y confiabilidad del modelo, crucial en aplicaciones como diagnósticos médicos o detección de fraudes financieros. La identificación y tratamiento adecuado de outliers son pasos críticos para construir modelos de machine learning precisos y confiables, fundamentales para la toma de decisiones precisas y seguras en diversas industrias y aplicaciones.

- **Estandarización de los datos**: Es una técnica utilizada para estandarizar el rango de variables independientes o predictoras. Su objetivo es estandarizar el rango de VP de los datos para evitar que alguna característica domine a las demás, especialmente al tratar con conjuntos de datos grandes. Este es un paso crucial en el preprocesamiento de datos, especialmente para algoritmos sensibles al rango de los datos, como los modelos de aprendizaje profundo. Existen varias formas de lograr el escalamiento de datos, incluyendo la normalización Min-Max y la Estandarización. La normalización Min-Max escala los datos dentro de un rango fijo (generalmente de 0 a 1), mientras que la Estandarización escala los datos con una media de 0 y una desviación estándar de 1.

- **Codificación de Variables Categóricas**: Los modelos requieren que las variables sean numéricas. Si los datos contienen datos categóricos, debemos codificarlos a valores numéricos antes de ajustar y evaluar un modelo. Este proceso, conocido como codificación de variables categóricas, es una técnica común de preprocesamiento de datos. Un método común es la codificación One-Hot, que crea nuevas columnas binarias para cada categoría/etiqueta en las columnas originales.


- **Selección de variables**: La selección de características es un proceso en el aprendizaje automático donde selecciona automáticamente aquellas características en sus datos que más contribuyen a la variable de predicción o salida en la que está interesado. Tener características irrelevantes en sus datos puede disminuir la precisión de muchos modelos, especialmente algoritmos lineales como la regresión lineal y logística. Este proceso es particularmente importante para los científicos de datos que trabajan con datos de alta dimensionalidad, ya que reduce el sobreajuste, mejora la precisión y reduce el tiempo de entrenamiento. Tres de las ventajas de realizar la selección de características antes de modelar sus datos son:

  - Reduce el Sobreajuste: Menos datos redundantes significa menos oportunidades para tomar decisiones basadas en el ruido.
  - Mejora la Precisión: Menos datos engañosos significa que la precisión del modelado mejora.
  - Reduce el Tiempo de Entrenamiento: Menos puntos de datos reducen la complejidad del algoritmo y se entrena más rápido.

## Duplicados

Existen varias razones por las cuales pueden ocurrir duplicados en un conjunto de datos. Los principales motivos por lo que esto puede pasar son:

- Al ingresar datos manualmente, es posible que se introduzcan duplicados de forma accidental.

- Al unir datos de diferentes bases de datos o fuentes, puede haber registros que se repiten. Sin embargo, si no hay identificadores únicos consistentes (como un ID de cliente), los registros duplicados pueden no ser identificados correctamente.

- Los datos recopilados en diferentes momentos pueden capturar la misma información repetidamente.


Gestionar los valores duplicados es esencial en el preprocesamiento de datos, ya que los duplicados pueden sesgar los resultados y reducir la eficacia del modelo. Algunas estrategias comunes para manejar los duplicados:


- Elimina filas que son duplicadas exactas de otras.

- Si ciertas columnas pueden contener duplicados significativos mientras que otras no, elimina duplicados basados en un subconjunto de columnas.

- En lugar de eliminar duplicados, agrupa por las columnas duplicadas y aplica una función de agregación (por ejemplo, promedio, suma) a otras columnas.

- Mantén la primera o la última aparición de cada duplicado y elimina el resto.

- Si los duplicados tienen información complementaria, combina los datos antes de eliminar los duplicados.

```python
# lo primero que vamos a hacer es extraer los valores que son duplicados
df_duplicados = df[df.duplicated()]
df.head()
```
