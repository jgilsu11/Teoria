# Introducción a la Estadística Descriptiva

## ¿Qué es la Estadística Descriptiva?

La estadística descriptiva es una rama fundamental de la estadística que se ocupa de la recopilación, organización, resumen y presentación de datos de una manera que facilite su interpretación. Su objetivo principal es proporcionar una comprensión inicial y visual de las características esenciales de un conjunto de datos, lo que es crucial para cualquier análisis. En concreto, nos va a permitir

- Condensar grandes volúmenes de datos en medidas clave como la media, mediana, moda, varianza y desviación estándar, las cuales nos van a ofrecer una visión general de la distribución y tendencias en los datos.

- Facilita la creación de gráficos y visualizaciones efectivas, ayudandonos a identificar patrones, tendencias y *outliers*.

- Nos ayuda a detectar *outliers*, que pueden ser indicativos de errores o eventos inusuales que requieren atención.

- Nos permite investigar relaciones entre variables, como la correlación, para entender cómo interactúan diferentes factores.

- Nos proporciona una base objetiva para la toma de decisiones basadas en datos, evaluando la calidad y contexto de los mismos.

## Tipos de Medidas en Estadística Descriptiva

Las medidas en estadística descriptiva se dividen en tres categorías principales:

### Medidas de Centralización
Estas medidas describen el valor central o típico de un conjunto de datos. Las medidas de centralización que nos podemos encontrar son:

- **Media (Promedio)**: Representa el valor promedio de los datos.

- **Mediana**: Es el valor que se encuentra en el centro de los datos ordenados.

- **Moda**: Es el valor que aparece con mayor frecuencia en el conjunto de datos.

### Medidas de Dispersión

Estas medidas indican cuán dispersos o agrupados están los datos alrededor de la medida de centralización:

- **Varianza**: Mide la dispersión de los datos respecto a la media.

- **Desviación Estándar**: Es la raíz cuadrada de la varianza, proporcionando una medida de dispersión en la misma unidad que los datos.

- **Rango**: Es la diferencia entre el valor máximo y el valor mínimo.

### Medidas de Forma
Estas medidas describen la forma de la distribución de los datos:

- **Asimetría (*Skewness*)**: Indica si la distribución de los datos está sesgada hacia la izquierda o hacia la derecha.

- **Curtosis**: Mide la concentración de los datos en el centro de la distribución.

### Otras Medidas Importantes

- **Medidas de Correlación**: Cuantifican la relación entre dos o más variables:

   - **Coeficiente de Correlación de Pearson**: Para variables continuas.

   - **Coeficiente de Correlación de Spearman**: Para variables ordinales.

   - **Coeficiente de Contingencia**: Para variables categóricas.

- **Medidas de Resumen Categóricas**: Describen y resumen datos categóricos o cualitativos:

   - **Tablas de Frecuencias**: Resumen la distribución de categorías o grupos en los datos.

- **Medidas de Posición Relativa**: Determinan la posición de un valor en relación con otros valores en una distribución:

   - **Percentiles**: Valores que dividen los datos en 100 partes iguales.

   - **Cuartiles**: Valores que dividen los datos en 4 partes iguales.
   
   - **Deciles**: Valores que dividen los datos en 10 partes iguales.

# Medidas de Centralización

**Las medidas de centralización** son herramientas clave en estadística que nos van a permitir identificar el valor más representativo o típico dentro de un conjunto de datos. Estas medidas ayudan a entender el centro de la distribución de los datos, proporcionando una visión clara de dónde se concentran los valores. Como ya hemos dicho, tenemos tres tipos de medidas de centralización: 

- **Media (Promedio):** La media es el valor promedio de un conjunto de datos. Se calcula sumando todos los valores y dividiéndolos por el número de observaciones. Es útil cuando los datos son bastante homogéneos y no hay valores atípicos extremos.

    Por ejemplo, si se analizan los ingresos mensuales de un grupo de personas y se obtienen los valores $100, $120, $110 y $130, la media sería $115. Esto indica que, en promedio, las personas ganan $115 al mes.

- **Mediana:** La mediana es el valor central en un conjunto de datos ordenados. Si el número de observaciones es impar, la mediana es el valor en el centro. Si es par, es el promedio de los dos valores centrales. Es especialmente útil cuando los datos incluyen valores atípicos que pueden distorsionar la media.

   Por ejemplo, en el análisis de salarios de una empresa, si los salarios son $30,000, $40,000, $50,000, $60,000 y $1,000,000, la mediana sería $50,000. Este valor representa mejor el salario típico que la media, que estaría influenciada por el salario extremo de $1,000,000.

- **Moda:** La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Es útil para identificar el valor más común o frecuente en un conjunto de datos.

   Por ejemplo, en una encuesta sobre el tipo de coche más popular, si el modelo más frecuentemente mencionado es el Modelo A, entonces la moda es el Modelo A. Esto indica que el Modelo A es el coche más popular entre los encuestados.


Sin embargo algunas consideraciones que tenemos que tener en cuenta a la hora de trabajar con este tipo de medidas son: 

- **Robustez:** La mediana es menos sensible a *outliers* que la media, lo que la hace más robusta en conjuntos de datos con *outliers*.

- **Aplicabilidad:** La moda puede ser menos informativa en conjuntos de datos con muchas categorías o sin un valor predominante claro.

- **Contexto:** Elegir la medida de centralización adecuada depende del contexto y de las características específicas de los datos analizados.

## Media

Para calcular la media en un DataFrame, podemos utilizar el método `.mean()`. La sintaxis básica para usar este método es la siguiente:

```python
df['NombreColumna'].mean()
```

Donde:

- `df`: Representa el DataFrame en el que se quiere calcular la media.

- `'NombreColumna'`: Es el nombre de la columna para la cual se desea calcular la media.

La media es un valor central que se obtiene sumando todos los valores de una columna y dividiendo el resultado entre la cantidad de valores. Es una medida común en estadística para representar el promedio de un conjunto de datos.

**Ventajas que tiene usar la media**

- La media es fácil de calcular y entender, lo que la hace una herramienta accesible para análisis básicos.

- Es ampliamente utilizada en diversas disciplinas, como economía, sociología, y ciencias de la salud, lo que facilita la comparación y el intercambio de información.

- En distribuciones simétricas, la media proporciona un buen valor central que representa a todo el conjunto de datos.

**Inconvenientes de la media**

- La media puede ser significativamente afectada por valores atípicos o extremos, lo que puede distorsionar la representación del conjunto de datos.

- En distribuciones asimétricas, la media puede no reflejar adecuadamente la tendencia central del conjunto de datos.

- La media solo es aplicable a datos cuantitativos, por lo que no puede usarse para analizar datos categóricos o nominales.

- Al condensar todos los datos en un solo valor, se pierde información sobre la variabilidad y la distribución de los datos originales.

````python
# Calculamos la media de los ingresos de los clientes.
# Interpretación: La media de ingresos representa el valor promedio de ingresos en el conjunto de datos de clientes. En este contexto, nos permite entender cuál es el ingreso promedio de los clientes en la muestra.
media_ingresos = df['Income'].mean()
print(f"La media de los ingresos de los clientes es: {np.round(media_ingresos, 2)}")
````

````python
# Calculamos la media de edad de los clientes.
# Interpretación: La media de edad nos proporciona la edad promedio de los clientes en el conjunto de datos. Esto puede ser útil para tener una idea general de la edad de la población de clientes.

media_edad = df['Age'].mean()
print(f"La media de edad de los clientes es: {np.round(media_edad, 2)}")
````

## Mediana

Para calcular la mediana en un DataFrame, utilizamos el método `.median()`. La sintaxis básica para usar este método es la siguiente:

```python
df['NombreColumna'].median()
```

Donde:

- `df`: Representa el DataFrame en el que se quiere calcular la mediana.

- `'NombreColumna'`: Es el nombre de la columna para la cual se desea calcular la mediana.

La mediana es el valor central de un conjunto de datos ordenados. Si el número de observaciones es impar, la mediana es el valor en el medio. Si el número es par, es el promedio de los dos valores centrales. La mediana es especialmente útil en distribuciones asimétricas, ya que no se ve afectada por valores extremos.

**Ventajas de la mediana**

- A diferencia de la media, la mediana no se ve afectada por valores atípicos o extremos, proporcionando una representación más precisa de la tendencia central en distribuciones asimétricas.

- La mediana divide el conjunto de datos en dos mitades, lo que la hace intuitiva y fácil de interpretar.

- En conjuntos de datos con valores atípicos, la mediana permanece estable, evitando la distorsión que podría afectar a la media.

**Inconvenientes de la mediana**

- Al igual que la media, la mediana condensa la información del conjunto de datos en un solo valor, perdiendo detalles sobre la distribución de los datos.

- La mediana no refleja pequeños cambios en el conjunto de datos tan claramente como lo hace la media, lo que puede ser una desventaja en ciertos análisis.

````python
#  Calculamos la mediana de los ingresos de los clientes.
# Interpretación: La mediana de ingresos es el valor medio de ingresos en el conjunto de datos de clientes cuando los ingresos se ordenan en orden ascendente. En este contexto, nos proporciona el valor de ingresos que se encuentra en la mitad del conjunto de datos, lo que significa que el 50% de los clientes tienen ingresos por encima de este valor y el 50% por debajo.

mediana_ingresos = df['Income'].median()
print(f"La media de los ingresos de los clientes es: {np.round(media_ingresos, 2)}")
print(f"La mediana de los ingresos de los clientes es: {np.round(mediana_ingresos, 2)}")
````

En otras palabras, es como si distribuyéramos equitativamente el total de los ingresos entre todos los clientes, resultando en que cada uno recibiría alrededor de `53,719.58 $`.

Esto significa que la mitad de los clientes tiene ingresos menores o iguales a `50,596.5 $`, y la otra mitad tiene ingresos mayores o iguales a esa cantidad.

En resumen, la **media** es el promedio de todos los ingresos y puede verse afectada por valores extremos, mientras que la **mediana** es el ingreso que se encuentra justo en el medio cuando los ingresos se ordenan y no se ve influenciada por valores extremos. Ambas medidas son importantes para entender la distribución de los ingresos de los clientes.

## Moda

Para calcular la moda en un DataFrame, utilizamos el método `.mode()`. La sintaxis básica para usar este método es la siguiente:

```python
df['NombreColumna'].mode()
```

Donde:

- `df`: Representa el DataFrame en el que se quiere calcular la moda.

- `'NombreColumna'`: Es el nombre de la columna para la cual se desea calcular la moda.

La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Puede haber más de una moda si varios valores tienen la misma frecuencia máxima. La moda es especialmente útil para datos categóricos, donde identifica la categoría más común.

**Ventajas de la moda**

- La moda es fácil de entender y calcular, proporcionando una medida rápida de la tendencia central.

- A diferencia de la media y la mediana, la moda puede aplicarse a datos categóricos, lo que la hace versátil en el análisis de datos cualitativos.

- Al igual que la mediana, la moda no se ve afectada por valores atípicos, ya que se basa en la frecuencia de los valores.

- La moda identifica el valor o los valores más comunes en el conjunto de datos, ofreciendo una visión clara de las tendencias predominantes.


**Inconvenientes de la moda**

- Al igual que otras medidas de tendencia central, la moda reduce el conjunto de datos a un solo valor, lo que puede no representar adecuadamente la distribución de los datos.

- En conjuntos de datos con más de una moda (distribución multimodal), puede ser difícil interpretar los resultados y la moda puede perder su utilidad.

- En algunos casos, la moda puede no ser representativa del conjunto de datos, especialmente si la frecuencia del valor más común es baja.

- La moda es menos útil para datos continuos, donde los valores únicos pueden ser muy numerosos y la frecuencia máxima puede no ser significativa.

# Medidas de dispersión

**Las medidas de dispersión** son fundamentales en estadística, ya que nos permiten entender cuán extendidos o agrupados están los datos dentro de un conjunto. Mientras que las medidas de tendencia central (como la media, la mediana y la moda) nos indican el valor promedio de nuestros datos, las medidas de dispersión nos muestran cuánto se separan o dispersan estos datos alrededor de ese valor promedio.

Las tres medidas de dispersión más importantes que tenemos ahora son:

- **Varianza:** Mide cuán dispersos están los datos respecto a la media. Se calcula como la media de los cuadrados de las diferencias entre cada dato y la media. Una varianza alta indica que los datos están muy dispersos alrededor de la media, mientras que una varianza baja significa que los datos están más agrupados cerca de la media.

- **Desviación Estándar:** Es una medida comúnmente utilizada y más intuitiva. Es la raíz cuadrada de la varianza. Al igual que la varianza, la desviación estándar mide cuánto se alejan los datos individuales del valor promedio (media) del conjunto de datos.

- **Rango:** Es la diferencia entre el valor máximo y el valor mínimo en un conjunto de datos. El rango nos da una idea rápida de la dispersión total de los datos. Un rango amplio indica que los datos están muy dispersos, mientras que un rango estrecho sugiere que los datos están más agrupados.

Resumiendo:

| Concepto           | Definición                                                                                          | Ejemplo Aplicado al Análisis de Datos                                   | Interpretación                                                                                        |
|--------------------|-----------------------------------------------------------------------------------------------------|------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------|
| **Varianza**           | La varianza mide la dispersión o variabilidad de un conjunto de datos. Se calcula como la media de los cuadrados de las diferencias entre cada valor y la media. | Al analizar los tiempos de respuesta de un sitio web, la varianza indica cuánto varían los tiempos respecto a la media. | Una varianza alta en los tiempos de respuesta indica una gran variabilidad, lo que puede afectar negativamente la experiencia del usuario. |
| **Desviación Estándar**| La desviación estándar es la raíz cuadrada de la varianza y proporciona una medida de la dispersión en la misma unidad que los datos originales. | En la investigación de mercado, la desviación estándar de las edades de los consumidores puede indicar la dispersión de edades en un mercado objetivo. | Una desviación estándar baja significa que la mayoría de los consumidores tienen edades similares, mientras que una alta indica una mayor variabilidad en las edades. |
| **Rango**              | El rango es la diferencia entre el valor máximo y el valor mínimo en un conjunto de datos. | Al evaluar la variabilidad de los precios de un producto, el rango indica la diferencia entre el precio más alto y el más bajo registrado. | Un rango amplio sugiere una gran variabilidad en los precios, mientras que un rango estrecho indica precios más consistentes. |

- `'NombreColumna'`: Es el nombre de la columna para la cual se desea calcular la varianza.

La varianza mide la dispersión de los datos respecto a su media. Específicamente, es el promedio de las diferencias al cuadrado entre cada valor y la media del conjunto de datos. Una varianza alta indica que los datos están muy dispersos alrededor de la media, mientras que una varianza baja indica que están más concentrados.

**Ventajas de la varianza**

- La varianza proporciona una medida cuantitativa de la dispersión de los datos, lo que es esencial para comprender la variabilidad dentro del conjunto de datos.

La desviación estándar mide la dispersión de los datos respecto a su media. Es la raíz cuadrada de la varianza y se expresa en las mismas unidades que los datos originales. Una desviación estándar alta indica una gran dispersión de los datos alrededor de la media, mientras que una baja indica que los datos están más concentrados.

**Ventajas de la desviación estándar**

- A diferencia de la varianza, la desviación estándar se expresa en las mismas unidades que los datos originales, lo que facilita su interpretación.

## Rango

Para calcular el rango en un DataFrame de Pandas, necesitamos restar el valor mínimo de la columna del valor máximo. Aunque Pandas no tiene un método específico para el rango, podemos calcularlo utilizando los métodos `.max()` y `.min()`. La sintaxis básica es la siguiente:

```python
rango = df['NombreColumna'].max() - df['NombreColumna'].min()
```

Donde:

- `df`: Representa el DataFrame en el que se quiere calcular el rango.

- `'NombreColumna'`: Es el nombre de la columna para la cual se desea calcular el rango.

El rango es la diferencia entre el valor máximo y el mínimo en un conjunto de datos. Proporciona una medida simple de la dispersión de los datos.

**Ventajas del rango**

- El rango es fácil de calcular y entender, proporcionando una medida rápida de la dispersión de los datos.

- Ayuda a identificar rápidamente los valores extremos en el conjunto de datos, proporcionando una visión inmediata del intervalo de los valores.

- Puede ser usado con cualquier tipo de datos cuantitativos, lo que lo hace una medida versátil.

- La diferencia entre el valor máximo y el mínimo es fácil de interpretar, ofreciendo una medida clara del spread de los datos.

**Inconvenientes del rango**

- El rango es altamente influenciado por valores atípicos, lo que puede distorsionar la representación de la dispersión de los datos.

- El rango solo tiene en cuenta los valores máximo y mínimo, ignorando completamente la distribución de los valores intermedios.

- Proporciona una medida muy general de la dispersión, lo que puede no ser suficiente para análisis detallados o comparaciones entre diferentes conjuntos de datos.

- Comparado con otras medidas de dispersión como la varianza o la desviación estándar, el rango proporciona menos información sobre la variabilidad de los datos.

````python
# Calculamos el rango de la Edad:
# Interpretación: El rango de las edades mide la diferencia entre la edad máxima y la edad mínima de los clientes. Un rango más amplio indicaría una mayor variabilidad en las edades, sugiriendo que hay una gran diferencia entre los clientes más jóvenes y los más viejos. Por otro lado, un rango más estrecho indicaría que las edades de los clientes están más agrupadas y son más similares entre sí.
rango_edad = df['Age'].max() - df['Age'].min()
print(f"El rango de edades es: {rango_edad}")
````

# Medidas de forma

**Las medidas de forma** son esenciales en estadística para entender la distribución y la forma de los datos en un conjunto. Mientras que las medidas de tendencia central (como la media, la mediana y la moda) y las medidas de dispersión (como la varianza y la desviación estándar) nos indican el valor promedio y la dispersión de los datos, las medidas de forma nos ayudan a comprender la simetría y la "cola" de la distribución de los datos.

Las dos medidas de forma más importantes que tenemos son:

- **Coeficiente de Asimetría (Skewness):** Mide la simetría de la distribución de los datos. Un coeficiente de asimetría positivo indica que la cola derecha (mayores valores) es más larga o está más extendida que la cola izquierda. Un coeficiente negativo indica lo contrario. Un coeficiente de asimetría cercano a cero indica que los datos están distribuidos simétricamente.

- **Curtosis:** Mide la "cola" y el "pico" de la distribución en comparación con una distribución normal. Una curtosis alta indica que los datos tienen colas más largas y picos más altos (más concentración de datos en la media y en los extremos), mientras que una curtosis baja indica una distribución con colas más cortas y menos concentración en el centro.

Un valor positivo indica que la cola derecha es más larga, mientras que un valor negativo indica que la cola izquierda es más larga. Un valor cercano a cero indica una distribución simétrica. | Al analizar los ingresos de los clientes, un coeficiente de asimetría positivo indicaría que hay más clientes con ingresos superiores a la media, mientras que uno negativo indicaría lo contrario. | Un coeficiente de asimetría positivo sugiere una mayor cantidad de valores extremos en el lado derecho de la distribución, mientras que uno negativo sugiere más valores extremos en el lado izquierdo. |

## Asimetría
Para calcular la asimetría en un DataFrame de Pandas, utilizamos el método `.skew()`. La sintaxis básica para usar este método es la siguiente:

```python
df['NombreColumna'].skew()
```

Donde:

- `df`: Representa el DataFrame en el que se quiere calcular la asimetría.

- `'NombreColumna'`: Es el nombre de la columna para la cual se desea calcular la asimetría.

La asimetría mide la falta de simetría en la distribución de los datos. Una asimetría positiva indica que la cola derecha es más larga o más pesada que la izquierda, mientras que una asimetría negativa indica lo contrario. La asimetría es útil para entender la distribución de los datos en relación con la media.

**Ventajas de la asimetría**

- La asimetría proporciona información sobre la forma de la distribución de los datos, ayudando a identificar si los datos están sesgados a la derecha o a la izquierda.

- Es útil para detectar sesgos en los datos, lo que puede ser crucial en análisis estadísticos y modelado predictivo.

- La asimetría complementa otras medidas estadísticas, como la media y la desviación estándar, ofreciendo una visión más completa de la distribución de los datos.


**Inconvenientes de la asimetría**

- La interpretación de la asimetría puede ser menos intuitiva que otras medidas estadísticas, especialmente para quienes no están familiarizados con el concepto.

- La asimetría puede ser influenciada por valores atípicos, lo que puede distorsionar la percepción de la distribución de los datos.

- La asimetría calculada a partir de una muestra puede no representar adecuadamente la asimetría de la población completa, especialmente en muestras pequeñas.

- Al igual que otras medidas de dispersión, la asimetría solo es aplicable a datos cuantitativos y no puede utilizarse para analizar datos categóricos o cualitativos.

La asimetría es una medida valiosa en el análisis de datos, proporcionando una comprensión adicional de la distribución y sesgos presentes en los datos. Sin embargo, es importante considerar sus limitaciones y utilizarla junto con otras medidas estadísticas para obtener una visión más completa y precisa del comportamiento de los datos.

**¿Cómo interpretamos el valor que nos da `stats` de la asimetría?**

- **Asimetría Positiva (*Skewness* > 0)**: Indica que la cola derecha de la distribución es más larga o tiene más peso que la cola izquierda. La mayor parte de los valores se agrupan en el lado izquierdo del gráfico, pero existen valores atípicos (*outliers*) que extienden la cola derecha.

- **Asimetría Negativa (*Skewness* < 0)**: Indica que la cola izquierda de la distribución es más larga o tiene más peso que la cola derecha. La mayor parte de los valores se agrupan en el lado derecho del gráfico, pero existen valores atípicos que extienden la cola izquierda.

- **Asimetría Cero (Skewness = 0)**: Indica que la distribución es simétrica.Las colas a ambos lados del promedio son aproximadamente iguales en longitud y peso.

````python
# Calculamos la simetría de la edad de los clientes.
# Interpretación: La asimetría de 0.97 indica que la distribución de edades de los clientes está sesgada hacia la derecha. Esto significa que la mayoría de los clientes son relativamente jóvenes, pero hay algunos clientes significativamente mayores que aumentan la cola derecha de la distribución. Este valor positivo de asimetría sugiere una mayor presencia de edades más altas entre los clientes en comparación con la media. 
asimetria_edad = stats.skew(df["Age"])
print(f"Asimetría de la variable edad es {round(asimetria_edad, 2)}")
````

La curtosis mide la "agudeza" de la distribución de los datos. Indica la presencia de valores extremos y la forma de la distribución en comparación con una distribución normal. Una curtosis alta indica una distribución con colas más pesadas y picos más altos, mientras que una curtosis baja indica una distribución con colas más ligeras y picos más bajos.

**Ventajas de la curtosis**

- Proporciona información sobre la forma de la distribución en términos de su "apuntamiento" o "planitud", complementando la información proporcionada por la asimetría.

- Es especialmente útil en finanzas y análisis de riesgos, donde la identificación de eventos extremos (como grandes pérdidas o ganancias) es importante.

- La curtosis, junto con la media, desviación estándar y asimetría, ofrece una visión completa de la distribución de los datos, mejorando la precisión de los análisis estadísticos.

**Inconvenientes de la curtosis**

- La interpretación de la curtosis puede ser menos intuitiva que otras medidas estadísticas, especialmente para aquellos sin formación en estadística.

- Si bien esto puede ser una ventaja, también puede ser una desventaja, ya que la presencia de unos pocos valores extremos puede influir significativamente en la curtosis, distorsionando la percepción general de la distribución.

- La curtosis calculada a partir de una muestra puede no representar adecuadamente la curtosis de la población completa, especialmente en muestras pequeñas.

- Al igual que otras medidas de dispersión y forma, la curtosis solo es aplicable a datos cuantitativos y no puede utilizarse para analizar datos categóricos o cualitativos.

La curtosis es una medida que describe la forma de las colas de una distribución en comparación con una distribución normal. Se centra en la extremidad y la dispersión de la distribución. Los valores de la curtosis pueden ser positivos, negativos o cercanos a cero, y se interpretan de la siguiente manera:

**¿Cómo interpretamos los resultados?**

- **Curtosis positiva (leptocúrtica) curtosis > 0**: Indica que la distribución tiene colas más pesadas y un pico más alto y estrecho en comparación con una distribución normal.  

- **Curtosis cercana a cero (mesocúrtica) curtosis ≈ 0**: Indica que la distribución tiene una forma similar a la distribución normal, con colas y pico moderados.

- **Curtosis negativa (platicúrtica) exceso de curtosis < 0**: Indica que la distribución tiene colas más ligeras y un pico más bajo y ancho en comparación con una distribución normal.

## Medidas de Correlación

**Las medidas de correlación** son utilizadas para entender cómo dos variables están relacionadas entre sí. Nos ayudan a descubrir si existe alguna conexión o patrón entre las variables que estamos estudiando.

**¿Por qué son importantes?**

Imagina que estamos analizando datos de ventas de helados y la temperatura en diferentes días. Queremos saber si la temperatura afecta las ventas de helados. Este tipo de medidas nos van a permitir contestar preguntas del tipo: "¿Cuando hace más calor, se venden más helados?" o "¿Hay alguna relación entre estas dos variables?".


**¿Cómo se mide la correlación?**

Una de las medidas más comunes de correlación es el "coeficiente de correlación", representado con la letra "r". Este coeficiente puede tomar valores entre -1 y 1:

- **Si r es cercano a 1**: Indica una correlación positiva fuerte. Esto significa que cuando una variable aumenta, la otra también tiende a aumentar. Por ejemplo, si la temperatura sube, es más probable que las ventas de helados aumenten. En este caso, la correlación es positiva. 

- **Si r es cercano a -1**: Indica una correlación negativa fuerte. Esto significa que cuando una variable sube, la otra tiende a bajar de manera consistente. Cuando una variable aumenta, la otra tiende a disminuir. Por ejemplo, si el precio de un producto sube, es probable que la cantidad de personas que lo compran disminuya. Aquí, la correlación es negativa.

- **Si r está cerca de 0**: Indica que no hay una relación clara entre las dos variables. Esto significa que un cambio en una variable no implica ningún cambio en la otra. Por ejemplo, la cantidad de libros leídos en un año y la cantidad de café consumido pueden no tener ninguna relación directa, mostrando una ausencia de correlación.

Para calcular la correlación, usaremos el método `.corr()`. Aquí está la sintaxis básica:

```python
correlation_matrix = df.corr(method='pearson', min_periods=1, numeric_only=True)
```

Donde:

- `df`: Es el DataFrame con el que deseamos calcular las correlaciones. Devuelve una matriz de correlación que muestra la relación entre todas las columnas numéricas del DataFrame.

- `method`: Este parámetro nos permite especificar el método de cálculo de la correlación. Los valores más usuales son:

  - `'pearson'` (por defecto): Calcula la correlación de Pearson, que mide la relación lineal entre dos variables continuas.

  - `'kendall'`: Calcula la correlación de Kendall, que mide la correlación de rango entre dos variables.

  - `'spearman'`: Calcula la correlación de Spearman, que es una versión no paramétrica de la correlación de Pearson basada en los rangos de los datos.

- `min_periods`: Establece un número mínimo de observaciones requeridas para calcular una correlación. Por defecto, es `1`.

- `numeric_only`: Este parámetro especifica que la matriz de correlación se haga sobre variables numéricas.

### Métodos de Correlación

- **Correlación de Pearson**: Mide la relación lineal entre dos variables. Detecta **únicamente relaciones lineales**, es decir, cuando el cambio en una variable se relaciona directamente con el cambio en la otra de manera constante. Si embargo, este método es muy sensible a *outliers* que pueden distorsionar los resultados.

- **Correlación de Spearman**: Detecta relaciones monotónicas, es decir, cuando las variables tienden a aumentar o disminuir juntas, pero no necesariamente de manera constante. A menudo se utiliza cuando se quiere evaluar la relación entre dos variables ordinales (datos que se pueden ordenar pero no necesariamente tienen una escala numérica) o cuando se sospecha que la **relación es no lineal**. Se usa en situaciones donde los valores de las variables no tienen un significado numérico directo, como la calificación de películas por parte de espectadores.

- **Correlación de Kendall**: Detecta relaciones monotónicas, al igual que la correlación de Spearman. Robusta frente a valores atípicos y adecuada cuando no se conoce la forma específica de la relación entre las variables. Se utiliza en situaciones donde se quiere evaluar si hay una relación de orden entre variables, como la clasificación de productos según la preferencia de los consumidores.

sns.heatmap(data, cmap, annot, fmt, linewidths, linecolor, cbar, ax, vmax, vmin)

Donde:

- **`data`**: Es el parámetro más importante y representa los datos que deseas visualizar en el heatmap. Puede ser una matriz bidimensional (por ejemplo, un DataFrame de pandas) que contenga los valores que se mostrarán en el heatmap.

- **`cmap`** (opcional): Controla la paleta de colores que se utilizará en el heatmap. Puedes proporcionar el nombre de un esquema de colores predefinido o una paleta de colores personalizada. Ejemplos comunes incluyen "viridis", "coolwarm", "YlGnBu", entre otros. Seaborn tiene muchas paletas integradas que puedes explorar.

- **`annot`** (opcional): Este parámetro es un booleano (True o False) que determina si se mostrarán los valores en las celdas del heatmap. Si se establece en True, los valores se mostrarán en las celdas; si se establece en False (valor predeterminado), no se mostrarán.

- **`fmt`** (opcional): Define el formato de los valores en las celdas si `annot` está configurado en True. Puedes especificar un formato utilizando cadenas de formato de estilo Python. Por ejemplo, `fmt=".2f"` mostrará los valores con dos decimales.

- **`linewidths`** (opcional): Controla el ancho de las líneas que separan las celdas del heatmap. Puedes establecerlo en un valor numérico para personalizar el ancho de las líneas. Si se establece en 0 (valor predeterminado), no se mostrarán líneas entre las celdas.

- **`linecolor`** (opcional): Define el color de las líneas que separan las celdas si `linewidths` es mayor que 0. Puedes especificar un color utilizando un nombre de color o un código hexadecimal.

- **`cbar`** (opcional): Es un booleano que controla si se muestra la barra de color (colorbar) junto al heatmap para indicar la escala de colores. Si se establece en True (valor predeterminado), se mostrará; si se establece en False, no se mostrará.

- **`ax`** (opcional): Este parámetro te permite especificar el objeto de ejes Matplotlib en el que se dibujará el heatmap. Si no se proporciona, Seaborn creará automáticamente un nuevo objeto de ejes.

- **`vmin`** (opcional): Especifica el valor mínimo en la escala de colores. Los valores en el heatmap que sean iguales o menores que `vmin` se asignarán al color correspondiente al valor mínimo de la paleta de colores (`cmap`). Todo lo que esté por debajo de `vmin` se mostrará con el color mínimo de la paleta.

- **`vmax`** (opcional): Especifica el valor máximo en la escala de colores. Los valores en el heatmap que sean iguales o mayores que `vmax` se asignarán al color correspondiente al valor máximo de la paleta de colores (`cmap`). Todo lo que esté por encima de `vmax` se mostrará con el color máximo de la paleta.

- **`mask`** (opcional): Se utiliza para ocultar ciertas celdas en un heatmap o en otras visualizaciones de datos basadas en matrices, como los mapas de correlación.

pd.crosstab(df['Columna1'], df['Columna2'])

- Ayudan a detectar dependencias o independencias entre variables categóricas, lo que puede ser crucial en análisis estadísticos y estudios de investigación.

## Medidas de Posición Relativa

**Las medidas de posición relativa** son esenciales en estadística para entender cómo se sitúan los datos en relación con otros datos en un conjunto. Mientras que las medidas de tendencia central (como la media y la mediana) y las medidas de dispersión (como la varianza y la desviación estándar) nos indican el valor promedio y la dispersión de los datos, las medidas de posición relativa nos ayudan a comprender la posición de un dato específico dentro del conjunto de datos.

Las tres medidas de posición relativa más importantes son:

- **Cuartiles:** Dividen el conjunto de datos en cuatro partes iguales. Los cuartiles ayudan a entender la dispersión y la distribución de los datos.

- **Percentiles:** Dividen el conjunto de datos en cien partes iguales. Los percentiles son útiles para comparar datos individuales con el conjunto completo de datos.

- **Deciles:** Dividen el conjunto de datos en diez partes iguales. Los deciles proporcionan una visión más detallada de la distribución de los datos que los cuartiles, pero menos detallada que los percentiles.

Resumiendo:

| Medida de Posición Relativa | Definición                                                                                          | Ejemplo Aplicado al Análisis de Datos                                   | Interpretación                                                                                        |
|-----------------------------|-----------------------------------------------------------------------------------------------------|------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------|
| **Cuartiles**               | Dividen el conjunto de datos en cuatro partes iguales. | Al analizar los salarios de empleados, los cuartiles pueden mostrar cómo se distribuyen los salarios, con el 25% más bajo en el primer cuartil y el 25% más alto en el cuarto cuartil. | Ayudan a entender la dispersión de los datos y a identificar outliers en cada cuarto del conjunto de datos. |
| **Percentiles**             | Dividen el conjunto de datos en cien partes iguales. | En un examen estandarizado, si un estudiante está en el percentil 90, significa que el 90% de los estudiantes obtuvieron una puntuación igual o menor. | Permiten comparar la posición relativa de un dato individual en el contexto del conjunto completo de datos. |
| **Deciles**                 | Dividen el conjunto de datos en diez partes iguales. | Al evaluar las puntuaciones de un test, los deciles pueden mostrar cómo se distribuyen las puntuaciones en diez grupos, cada uno con el 10% de los datos. | Ofrecen una visión detallada de la distribución de los datos, identificando la posición de los datos dentro de los diez segmentos. |


### Cuartiles

Los cuartiles son una herramienta estadística esencial para dividir un conjunto de datos en cuatro partes iguales, cada una representando un cuarto de la población muestral. Se utilizan para entender la dispersión y la distribución de los datos. Los cuartiles son particularmente útiles en el análisis descriptivo, permitiendo identificar la concentración y la variabilidad de los datos.

En Pandas, se pueden calcular fácilmente utilizando el método `quantile()`. La sintaxis básica para calcular los cuartiles es la siguiente:

```python
Q1 = df['NombreColumna'].quantile(0.25)
Q2 = df['NombreColumna'].quantile(0.50)  # También conocido como la mediana
Q3 = df['NombreColumna'].quantile(0.75)

# Si la columna contiene valores NaN
Q1_nan = df['NombreColumna'].nanquantile(0.25)
Q2_nan = df['NombreColumna'].nanquantile(0.50)
Q3_nan = df['NombreColumna'].nanquantile(0.75)
```

Donde:

- `df`: Representa el DataFrame en el que se quiere calcular los cuartiles.

- `'NombreColumna'`: Es el nombre de la columna para la cual se desean calcular los cuartiles.

**Ventajas de los cuartiles**

- Dividen los datos en cuatro partes iguales, facilitando el análisis de la dispersión y la distribución.

- Ayudan a identificar valores atípicos, ya que los datos que caen fuera del rango intercuartílico (Q1 - Q3) pueden ser considerados *outliers*.

- Ofrecen una visualización clara de cómo se distribuyen los datos, proporcionando una comprensión rápida de la concentración y variabilidad. 

**Inconvenientes de los cuartiles**

- Proporcionan una visión segmentada de los datos, pero no ofrecen detalles sobre la distribución interna dentro de cada cuartil.

- En conjuntos de datos pequeños, los cuartiles pueden no ser representativos de la población general.

- Requieren una interpretación cuidadosa, ya que los cuartiles extremos pueden ser influenciados por valores atípicos.

- Al igual que otras medidas estadísticas de dispersión, los cuartiles solo son aplicables a datos cuantitativos y no pueden utilizarse para analizar datos categóricos o cualitativos.

````python
# Calculamos los percentiles 25 y 75 de la Columna 'age'
cuartil_25 = np.nanquantile(df["Age"], 0.25)
cuartil_75 = np.nanquantile(df["Age"], 0.75)

# mostramos los resultados
print(f'El cuartil 25 de la columna "Age" es: {cuartil_25}')
print(f'El cuartil 75 de la columna "Age" es: {cuartil_75}')

# Interpretación
# El cuartil 25 (primer cuartil) nos indica que el 25% de los clientes tienen una edad igual o menor que 24.
# El cuartil 75 (tercer cuartil) nos indica que el 75% de clientes tienen una edad igual o menor a 33. 
````

### Percentiles

Los percentiles dividen un conjunto de datos en cien partes iguales, permitiendo comprender la posición relativa de un dato específico dentro del conjunto. Se utilizan para comparar datos individuales con el conjunto completo, facilitando la identificación de la distribución y dispersión de los datos.

En NumPy, se pueden calcular fácilmente utilizando los métodos `percentile()` y `nanpercentile()` (este último es útil cuando el conjunto de datos contiene valores NaN que se deben ignorar). Además, estos métodos permiten pasar una lista de los percentiles que queremos calcular. La sintaxis básica para calcular los percentiles es la siguiente:

```python
import numpy as np

# Calcular percentiles individuales
P90 = np.percentile(df['NombreColumna'], 90)
P50 = np.percentile(df['NombreColumna'], 50)  # También conocido como la mediana
P10 = np.percentile(df['NombreColumna'], 10)

# Calcular una lista de percentiles
percentiles = np.percentile(df['NombreColumna'], [10, 25, 50, 75, 90])

# Si la columna contiene valores NaN
P90_nan = np.nanpercentile(df['NombreColumna'], 90)
P50_nan = np.nanpercentile(df['NombreColumna'], 50)
P10_nan = np.nanpercentile(df['NombreColumna'], 10)

# Calcular una lista de percentiles ignorando NaNs
percentiles_nan = np.nanpercentile(df['NombreColumna'], [10, 25, 50, 75, 90])
```

Donde:

- `df`: Representa el DataFrame en el que se quiere calcular los percentiles.

- `'NombreColumna'`: Es el nombre de la columna para la cual se desean calcular los percentiles.

- `percentiles`: Es una lista de los percentiles que queremos calcular.

**Ventajas de los percentiles**

- Permiten comparar un dato individual con el conjunto completo, facilitando la identificación de su posición relativa.

- Ayudan a identificar valores atípicos y extremos en los datos.

- Proporcionan una visualización clara de cómo se distribuyen los datos en diferentes segmentos, facilitando la interpretación y el análisis de la dispersión.


**Inconvenientes de los percentiles**

- La interpretación de percentiles puede ser menos intuitiva, especialmente para quienes no están familiarizados con estos conceptos.

- La utilidad de los percentiles puede verse afectada por la distribución de los datos, especialmente en distribuciones no normales.

- Indican la posición relativa de un dato, pero no proporcionan información sobre la magnitud de las diferencias entre los datos.

- Al igual que otras medidas estadísticas, los percentiles solo son aplicables a datos cuantitativos y no pueden utilizarse para analizar datos categóricos o cualitativos.

````python
# Calculamos los percentiles 25 y 75 de la Columna 'age'
percentile_25_75 = np.nanpercentile(df["Age"], [25, 75])

# mostramos los resultados
print(f'El percentil 25 de la columna "Age" es: {percentile_25_75[0]}')
print(f'El percentil 75 de la columna "Age" es: {percentile_25_75[1]}')

# Interpretación
# El percentil 25 (primer cuartil) nos indica que el 25% de los clientes tienen una edad igual o menor que 24.
# El percentil 75 (tercer cuartil) nos indica que el 75% de clientes tienen una edad igual o menor a 33. 
````
